## 제목 전처리

```python
import requests

url = "https://raw.githubusercontent.com/stopwords-iso/stopwords-ko/master/stopwords-ko.txt"
STOPWORDS = requests.get(url).text.splitlines()
```

```python
import re, numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer

# 0) 입력 가정
# df_all: 위 단계에서 만든 결합 DF (필수: 'title')
# 필요하면 중복/결측 제거
df_all = df_all.copy()
df_all = df_all.dropna(subset=["title"])
df_all = df_all.drop_duplicates(subset=["title"]).reset_index(drop=True)

# 1) 간단 토크나이저 + 불용어
def tokenize_ko(text: str):
    text = re.sub(r"[^0-9가-힣A-Za-z]+", " ", str(text))
    toks = text.lower().split()
    return [t for t in toks if len(t) > 1]

def analyzer(doc): return [t for t in tokenize_ko(doc) if t not in STOPWORDS]

df_all["tokens"] = df_all["title"].apply(analyzer)
print(df_all[["title", "tokens"]].head(10))
```

```python
# 2) TF-IDF 적합
vectorizer = TfidfVectorizer(
    analyzer=analyzer,
    ngram_range=(1,2),  # 단어+바이그램
    min_df=2            # 2개 이상 문서에서 등장한 토큰만
)
X = vectorizer.fit_transform(df_all["title"].astype(str))
vocab = np.array(vectorizer.get_feature_names_out())

# 3) 전체 상위 키워드(top-K)
top_k = 100
scores = np.asarray(X.sum(axis=0)).ravel()
order = scores.argsort()[::-1][:top_k]
top_keywords = list(zip(vocab[order], scores[order]))

print("[전체 상위 키워드]")
for k, s in top_keywords:
    print(f"{k}\t{s:.3f}")
```

## 키워드 추출

```python
# =========================
# YouTube 제목 키워드/템플릿 추출 (한 셀 완결)
# =========================

# 0) Imports
import os, re, requests, numpy as np, pandas as pd
from tqdm import tqdm

# 1) 입력 정리 (title 필수)
assert "df_all" in globals(), "df_all DataFrame이 메모리에 있어야 합니다."
assert "title" in df_all.columns, "df_all에 'title' 컬럼이 필요합니다."
df_all = df_all.copy().dropna(subset=["title"]).drop_duplicates(subset=["title"]).reset_index(drop=True)

# 2) 한국어 불용어 세트 로드(깃허브 + 폴백) — 반드시 set 형태로!
try:
    url = "https://raw.githubusercontent.com/stopwords-iso/stopwords-ko/master/stopwords-ko.txt"
    STOPWORDS = set(w.strip().lower() for w in requests.get(url, timeout=10).text.splitlines() if w.strip())
except Exception:
    STOPWORDS = {"그리고","그러나","하지만","그래서","또한","합니다","했다","하는","하게","되어","되는",
                 "에게","에서","으로","보다","같은","이번","오늘","그것","저것","이것","추천","후기","리뷰",
                 "모음","총정리","꿀팁","가이드","방법","비법","브이로그","영상","최신","필수","주의","실화","이슈"}
# placeholder류 추가
STOPWORDS |= {"url","email","hashtag","num","eng"}

# 3) 토크나이저(간단) + 불용어 적용
def tokenize_ko(text: str):
    text = re.sub(r"https?://\S+|www\.\S+", " url ", str(text))
    text = re.sub(r"[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}", " email ", text)
    text = re.sub(r"#\S+", " hashtag ", text)
    text = re.sub(r"\d+([.,]\d+)*", " num ", text)
    text = re.sub(r"[^0-9가-힣A-Za-z]+", " ", text)
    toks = text.lower().split()
    return [t for t in toks if len(t) > 1 and t not in STOPWORDS]


df_all["tokens"] = df_all["tokens"].apply(
    lambda xs: [t.lower() for t in (xs or []) if isinstance(t, str) and t.strip() and t.lower() not in STOPWORDS]
)

print(df_all[["title","tokens"]].head(10))
```

```python
# 4) TF-IDF 키워드 (프리토큰 입력 전용 — stop_words 전달 금지)
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import normalize

def tfidf_keywords_from_tokens(tokens_list, top_k=6, ngram_range=(1,2), min_df=2, max_df=0.95):
    vec = TfidfVectorizer(
        analyzer=lambda x: x,   # 이미 토큰 리스트
        tokenizer=None,
        preprocessor=None,
        token_pattern=None,
        lowercase=False,
        ngram_range=ngram_range,
        min_df=min_df,
        max_df=max_df,
        stop_words=None         # ⚠️ 반드시 None (경고 방지)
    )
    X = vec.fit_transform(tokens_list)
    X = normalize(X, norm="l2", copy=False)
    vocab = vec.get_feature_names_out()

    results = []
    for i in tqdm(range(X.shape[0]), desc="TF-IDF keywords"):
        row = X.getrow(i)
        if row.nnz == 0:
            results.append([]); continue
        ord_idx = row.indices[row.data.argsort()[::-1][:max(top_k*6, top_k)]]
        cand = [(vocab[j], float(row[0, j])) for j in ord_idx]
        cleaned, added = [], set()
        for token, score in cand:
            if len(token) < 2: continue
            if token in added: continue
            cleaned.append((token, score)); added.add(token)
            if len(cleaned) >= top_k: break
        results.append(cleaned)
    return results, vec

tokens_list = df_all["tokens"].tolist()
tfidf_res, tfidf_vec = tfidf_keywords_from_tokens(tokens_list, top_k=6, ngram_range=(1,2), min_df=2, max_df=0.95)

# 5) KeyBERT 키워드 (한국어 불용어를 tokenizer에서 재적용해 경고 없이 일관 처리)
def keybert_from_tokens(tokens_list, top_k=6, model_name="snunlp/KR-SBERT-V40K-klueNLI-augSTS"):
    try:
        from keybert import KeyBERT
        from sentence_transformers import SentenceTransformer
        from sklearn.feature_extraction.text import CountVectorizer
    except Exception:
        return [[] for _ in tokens_list]

    # SBERT 로드(폴백 포함)
    try:
        model = SentenceTransformer(model_name)
    except Exception:
        model = SentenceTransformer("paraphrase-multilingual-MiniLM-L12-v2")
    kw_model = KeyBERT(model=model)

    # 커스텀 tokenizer: split → STOPWORDS 제거 (한국어 리스트 강제 적용)
    def _tok(s: str):
        words = s.split()
        return [w for w in words if w not in STOPWORDS and len(w) > 1]

    vec = CountVectorizer(
        tokenizer=_tok, preprocessor=None, token_pattern=None,
        lowercase=False, ngram_range=(1,3), stop_words=None, min_df=1
    )

    results = []
    for toks in tqdm(tokens_list, desc="KeyBERT keywords"):
        text = " ".join(toks) if isinstance(toks, list) else str(toks)
        kws = kw_model.extract_keywords(
            text,
            vectorizer=vec,        # 후보 생성에 우리 tokenizer 사용
            use_maxsum=True,
            nr_candidates=20,
            top_n=top_k,
        )
        results.append([(k, float(s)) for k, s in kws])
    return results

keybert_res = keybert_from_tokens(tokens_list, top_k=6)


df_all = df_all.copy()
df_all["tfidf_keywords"] = tfidf_res
df_all["keybert_keywords"] = keybert_res
```

```python
# =========================
# 템플릿 생성 (word-level, 표면형 보존)
# 순서: Kkma로 고유명사/숫자 판단 → PERSON(보수) → NUM → ENG → LOC → NOUN → 괄호{X}
# =========================
import re, unicodedata
from tqdm import tqdm
from konlpy.tag import Kkma

kkma = Kkma()

# (있으면 사용; 없으면 빈 세트)
keep_tokens = {*(keep_tokens if 'keep_tokens' in globals() else set())}
STOPWORDS  = {*(STOPWORDS  if 'STOPWORDS'  in globals() else set())}

# ------ 유틸: 유니코드 안정화 & 토큰화(표면형 유지) ------
_SURROGATE_RE = re.compile(r"[\ud800-\udfff]")
_ZERO_WIDTH_RE = re.compile(r"[\u200B-\u200F\uFEFF]")
_EMOJI_SUPP_RE = re.compile(r"[\U00010000-\U0010FFFF]")

def clean_text(s: str) -> str:
    s = unicodedata.normalize("NFC", str(s))
    s = _ZERO_WIDTH_RE.sub(" ", s)
    s = _SURROGATE_RE.sub(" ", s)
    s = _EMOJI_SUPP_RE.sub(" ", s)
    return s

# 단어/구두점/공백을 분리하되, 원래 표면형과 순서를 그대로 보존
# ex) ["노홍철", " ", "또", " ", "한번", " ", "놀라게", " ", "만든", " ", "충격적인", " ", "동유럽", " ", "물가", "(", "불가리아", ")"]
WORD_SPLIT = re.compile(r"([^\w가-힣]+)")  # '단어'가 아닌 부분(스페이스/구두점)을 따로 캡처

def split_preserve(text: str):
    parts = []
    for chunk in WORD_SPLIT.split(text):
        if chunk is None or chunk == "":
            continue
        parts.append(chunk)
    return parts

def is_word(token: str) -> bool:
    # 한글, 영문, 숫자 포함된 묶음을 '단어'로 간주 (공백/구두점은 False)
    return bool(re.search(r"[A-Za-z0-9가-힣]", token))

# ------ 패턴 & 사전 ------
COMMON_SURNAMES = {
    "김","이","박","최","정","강","조","윤","장","임","한","오","서","신","권","황","안","송","전","홍","유","고","문",
    "양","손","배","백","허","남","심","노","하","곽","성","차","주","우","구","민","류","나","진","지","엄","채","원",
    "천","방","공","현","함","변","염","여","추","도","소","석","선","설","마","길","연","위","표","명","기","반","라",
    "왕","금","옥","육","인","맹","제","모","남궁","탁","국","어"
}
ENG_WORD  = re.compile(r"^[A-Za-z]{2,}$")           # 순수 영문(2+)
ENG_CAP   = re.compile(r"^[A-Z][a-z]{1,20}$")       # 영문 이름 토큰 모양
NUM_FULL  = re.compile(r"^\d+(?:[.,]\d+)*$")        # 순수 숫자
NUM_INNER = re.compile(r"\d+")                      # 혼합 숫자 포함(4K, 1080p 등)
BRACKETS_RE = re.compile(r"(\([^)]+\)|\[[^\]]+\])") # (), [] 만 {X} 후보

def mask_brackets_with_X(s: str, numeric_ratio=0.6):
    def _f(m):
        inner = m.group(0)[1:-1]
        if not inner.strip(): return m.group(0)
        digits = len(re.findall(r"\d", inner))
        nonalpha = len(re.findall(r"[^가-힣A-Za-z]", inner))
        total = max(1, len(inner))
        ratio = max(digits, nonalpha) / total
        if ratio >= numeric_ratio:
            return m.group(0)[0] + " {X} " + m.group(0)[-1]
        return m.group(0)
    return BRACKETS_RE.sub(_f, s)

# ------ 한국 전국 지역 사전 ------
# ① 17개 시·도 (광역단위)
LOC_KO = {
    # 특별시/광역시/특별자치시
    "서울","부산","대구","인천","광주","대전","울산","세종",
    # 도/특별자치도
    "경기","강원","충북","충남","전북","전남","경북","경남","제주","제주도",
    # 시급(주요 대도시 및 광역시 산하 구·시 일부)
    "수원","성남","용인","고양","과천","광명","안양","부천","안산","평택","의정부","김포","파주","남양주","하남","구리",
    "시흥","군포","의왕","오산","이천","여주","양평","동두천","가평","연천","포천",
    "춘천","원주","강릉","동해","속초","삼척","태백","홍천","횡성","영월","평창","정선","철원","화천","양구","인제","고성","양양",
    "청주","충주","제천","보은","옥천","영동","증평","진천","괴산","음성","단양",
    "천안","공주","보령","아산","서산","논산","계룡","당진","금산","부여","서천","청양","홍성","예산","태안",
    "전주","군산","익산","정읍","남원","김제","완주","진안","무주","장수","임실","순창","고창","부안",
    "목포","여수","순천","나주","광양","담양","곡성","구례","고흥","보성","화순","장흥","강진","해남","영암","무안","함평","영광","장성","완도","진도","신안",
    "포항","경주","김천","안동","구미","영주","영천","상주","문경","경산","군위","의성","청송","영양","영덕","청도","고령","성주","칠곡","예천","봉화","울진","울릉",
    "창원","진주","통영","사천","김해","밀양","거제","양산","의령","함안","창녕","고성","남해","하동","산청","함양","거창","합천",
    "제주시","서귀포시",
    # 광역시 산하 구·군(대표/자주 등장)
    "해운대","수영","남구","북구","동구","서구","중구","영도","금정","연제","동래","사하","사상","강서","기장",
    "수성","달서","달성","동구","서구","남구","북구","중구",
    "미추홀","연수","남동","부평","계양","강화","옹진",
    "남구","북구","광산",
    "유성","대덕",
    "남구","동구","중구","북구","울주",
    # 기타 많이 쓰는 관광도시/지역명
    "여의도","강남","홍대","명동","동대문","남대문","이태원","서촌","북촌","가평","양양","속초","여수","제주도","한라산","성산","중문","애월","협재"
}

# ------ PERSON 탐지(보수): 단어 시퀀스 기준 ------
def detect_person_spans(words, poss):
    """
    words: 단어/구두점/공백 전체 시퀀스
    poss:  각 '단어' 토큰에 대한 Kkma.pos 결과의 요약(label per word).
           비단어(공백/구두점)에는 None.
    반환: set[int]  -> PERSON으로 치환할 '단어' 인덱스 집합 (병합 시 첫 단어만 치환)
    규칙(보수):
      - 한글 이름: 연속 단어가 (1자/2자 + 2자) 등 2~4음절이고, 각 단어의 주요 품사가 NNP이며, 합친 문자열의 첫 글자가 성씨
      - 단일 단어 한글 2~3자, 주요 품사 NNP, 첫 글자 성씨
      - 영문 이름: ENG_CAP 단어 2개 연속 (둘 다 주요 품사 NNP로 나오지 않아도 허용)
    """
    person_heads = set()
    def is_nnp(i_word):
        labs = poss[i_word] or set()
        return ("NNP" in labs)
    def has_korean_only(s):
        return bool(re.fullmatch(r"[가-힣]+", s))
    # 1) 영문 풀네임
    for i in range(len(words)-1):
        if not (is_word(words[i]) and is_word(words[i+1])): 
            continue
        a, b = words[i], words[i+1]
        if ENG_CAP.match(a) and ENG_CAP.match(b):
            person_heads.add(i)  # i 위치에 {PERSON}을 넣고, i+1은 스킵
    # 2) 한글 이름 (공백 포함)
    n = len(words); i = 0
    while i < n:
        if not is_word(words[i]): 
            i += 1; continue
        w = words[i]
        if has_korean_only(w):
            # 단일 단어 2~3자, 성씨 시작 & NNP
            if 2 <= len(w) <= 3 and w[0] in COMMON_SURNAMES and is_nnp(i):
                person_heads.add(i); i += 1; continue
            # 공백 포함 2~4자
            j = i
            merged = w
            while j+1 < n and is_word(words[j+1]) and has_korean_only(words[j+1]) and len(merged) < 4:
                merged += words[j+1]; j += 1
            if 2 <= len(merged) <= 4 and merged[0] in COMMON_SURNAMES:
                ok = True
                for k in range(i, j+1):
                    labs = poss[k] or set()
                    if "NNP" not in labs: ok = False; break
                if ok:
                    person_heads.add(i)
                    i = j + 1
                    continue
        i += 1
    return person_heads

# ------ 각 단어에 대해 Kkma.pos 요약 ------
def pos_labels_per_word(token: str):
    """
    한 단어에 대해 Kkma.pos 실행 후, 주요 품사 set 반환.
    표면형 유지 목적이라 단어 단위로만 태깅/판단하고 치환은 단어 전체에 대해 수행.
    """
    labs = set()
    try:
        for t, p in kkma.pos(token):
            labs.add(p)
    except Exception:
        pass
    return labs

# ------ 메인: 한 제목 → 템플릿 ------
def to_template_surface(title: str,
                        min_noun_len=3,
                        max_mask_ratio=0.55,
                        max_person_per_title=2,
                        bracket_numeric_ratio=0.6):
    s0 = clean_text(title)
    # 0) URL/EMAIL/HASHTAG 선치환(표면형 보존)
    s = re.sub(r"https?://\S+|www\.\S+", " {URL} ", s0)
    s = re.sub(r"[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}", " {EMAIL} ", s)
    s = re.sub(r"#\S+", " {HASHTAG} ", s)

    # 1) 단어/구두점/공백 분리 (표면형 유지)
    parts = split_preserve(s)

    # 2) 각 '단어' 파트에 대해 Kkma.pos 요약 라벨 수집
    labels = [pos_labels_per_word(w) if is_word(w) and w not in {"{URL}","{EMAIL}","{HASHTAG}"} else None
              for w in parts]

    # 3) PERSON 헤드 인덱스 탐지(보수)
    person_heads = detect_person_spans(parts, labels)

    # 4) 단어별 치환: PERSON → NUM → ENG → LOC → NOUN (표면형 자체는 변형하지 않음)
    out = []
    person_used = 0
    masked = 0

    for i, w in enumerate(parts):
        # 비단어(공백/구두점) 또는 미리 치환된 플레이스홀더는 그대로
        if not is_word(w) or w in {"{URL}","{EMAIL}","{HASHTAG}"}:
            out.append(w); continue

        low = w.lower()

        # 4-1) PERSON (보수, 연속 span의 첫 단어에만 치환)
        if i in person_heads and person_used < max_person_per_title:
            out.append("{PERSON}")
            person_used += 1
            masked += 1
            continue

        # 4-2) 숫자: 순수/혼합
        labs = labels[i] or set()
        if NUM_FULL.match(w) or ("SN" in labs) or ("NR" in labs):
            out.append("{NUM}"); masked += 1; continue
        if NUM_INNER.search(w) and not ENG_WORD.match(w):
            out.append(NUM_INNER.sub("{NUM}", w)); masked += 1; continue

        # 4-3) 순수 ENG
        if ENG_WORD.match(w):
            out.append("{ENG}"); masked += 1; continue

        # 4-4) LOC (전국 지역 사전)  ← 추가된 부분
        if w in LOC_KO:
            out.append("{LOC}"); masked += 1; continue

        # 4-5) 상위 키워드/불용어 보호
        if low in keep_tokens or low in STOPWORDS:
            out.append(w); continue

        # 4-6) 일반명사/고유명사: 표면형을 {NOUN}으로만 치환(형태 변형 없음)
        if "NNP" in labs:
            out.append("{NOUN}"); masked += 1; continue
        if "NNG" in labs and len(w) >= min_noun_len:
            out.append("{NOUN}"); masked += 1; continue

        # 4-7) 그 외 품사(동사/형용사/조사 등)는 원형 유지
        out.append(w)

    # 5) 과마스킹 완화(비율 초과 시 일부 {NOUN}/{PERSON}/{LOC} 제거)
    total_tokens = max(1, sum(1 for w in parts if is_word(w)))
    if masked / total_tokens > max_mask_ratio:
        need = int(masked - max_mask_ratio * total_tokens)
        for j, t in enumerate(out):
            if need <= 0: break
            if t in ("{PERSON}","{NOUN}","{LOC}"):
                out[j] = ""  # 제거 → 공백 정리로 자연 축약
                need -= 1

    # 6) 조립 & 후처리 (연속 플레이스홀더 축약 + 괄호 {X})
    out_str = "".join(out)  # 표면형 보존: parts를 그대로 이어붙임
    out_str = re.sub(r"(?:\{PERSON\}\s*){2,}", "{PERSON} ", out_str)
    out_str = re.sub(r"(?:\{NUM\}\s*){2,}", "{NUM} ", out_str)
    out_str = re.sub(r"(?:\{NOUN\}\s*){2,}", "{NOUN} ", out_str)
    out_str = re.sub(r"(?:\{LOC\}\s*){2,}", "{LOC} ", out_str)
    out_str = re.sub(r"\s+", " ", out_str).strip()

    # 괄호 X 처리 (숫자/기호 위주일 때만)
    out_str = mask_brackets_with_X(out_str, numeric_ratio=bracket_numeric_ratio)

    return out_str

# ---------- 실행 ----------
templates = [to_template_surface(t,
                                 min_noun_len=3,
                                 max_mask_ratio=0.55,
                                 max_person_per_title=2,
                                 bracket_numeric_ratio=0.6)
             for t in tqdm(df_all["title"].astype(str), desc="Templates (surface-safe + LOC)")]

df_all["title_template"] = templates

template_summary = (
    df_all["title_template"].fillna("").replace("", None).dropna()
      .value_counts().reset_index().rename(columns={"index":"template","title_template":"count"})
)
print(template_summary.head(15))
```

```python
# =========================
# 템플릿 생성 (word-level, 표면형 보존)
# 순서: Kkma로 고유명사/숫자 판단 → PERSON → NUM(조사 보존) → ENG → LOC(조사 보존) → NOUN(조사 보존) → 괄호{X}
# =========================
import re, unicodedata
from tqdm import tqdm
from konlpy.tag import Kkma

kkma = Kkma()

# (있으면 사용; 없으면 빈 세트)
keep_tokens = {*(keep_tokens if 'keep_tokens' in globals() else set())}
STOPWORDS  = {*(STOPWORDS  if 'STOPWORDS'  in globals() else set())}

# ------ 유틸: 유니코드 안정화 & 토큰화(표면형 보존) ------
_SURROGATE_RE = re.compile(r"[\ud800-\udfff]")
_ZERO_WIDTH_RE = re.compile(r"[\u200B-\u200F\uFEFF]")
_EMOJI_SUPP_RE = re.compile(r"[\U00010000-\U0010FFFF]")

def clean_text(s: str) -> str:
    s = unicodedata.normalize("NFC", str(s))
    s = _ZERO_WIDTH_RE.sub(" ", s)
    s = _SURROGATE_RE.sub(" ", s)
    s = _EMOJI_SUPP_RE.sub(" ", s)
    return s

# 단어/구두점/공백 분리 (표면형 보존)
WORD_SPLIT = re.compile(r"([^\w가-힣]+)")

def split_preserve(text: str):
    parts = []
    for chunk in WORD_SPLIT.split(text):
        if chunk is None or chunk == "":
            continue
        parts.append(chunk)
    return parts

def is_word(token: str) -> bool:
    return bool(re.search(r"[A-Za-z0-9가-힣]", token))

# ------ 조사 사전(긴 것 우선) & 분리 ------
JOSA_SUFFIXES = [
    "으로써","으로서","으로부터","로부터",
    "이었습니다","였습니다","이었다","였다","이었다가","였다가",
    "이라고는","라고는","이라고","라고",
    "에게서","에게만","에게도","에게","께서","께",
    "에서는","에서","에는만","에는도","에는","에는도","에는",
    "으로는","로는","으로","로",
    "이라면","라면","이라서","라서","이지만","지만","이거나","거나",
    "이며서","면서","이며","며","이고","고",
    "과의","와의","과는","와는","과도","와도","과만","와만","과","와","랑은","랑도","랑만","랑",
    "은","는","이","가","을","를","도","만","뿐","까지","부터","마저","조차","이나","나","의"
]
JOSA_SUFFIXES = sorted(list({j for j in JOSA_SUFFIXES if re.fullmatch(r"[가-힣]+", j)}), key=len, reverse=True)

def split_stem_josa(word: str):
    """어절 끝에서 조사 하나를 떼어 (stem, josa)로 반환. 없으면 (word, "")"""
    for suf in JOSA_SUFFIXES:
        if word.endswith(suf) and len(word) > len(suf):
            return word[:-len(suf)], suf
    return word, ""

# ------ 패턴 & 사전 ------
COMMON_SURNAMES = {
    "김","이","박","최","정","강","조","윤","장","임","한","오","서","신","권","황","안","송","전","홍","유","고","문",
    "양","손","배","백","허","남","심","노","하","곽","성","차","주","우","구","민","류","나","진","지","엄","채","원",
    "천","방","공","현","함","변","염","여","추","도","소","석","선","설","마","길","연","위","표","명","기","반","라",
    "왕","금","옥","육","인","맹","제","모","남궁","탁","국","어"
}
ENG_WORD  = re.compile(r"^[A-Za-z]{2,}$")
ENG_CAP   = re.compile(r"^[A-Z][a-z]{1,20}$")
NUM_FULL  = re.compile(r"^\d+(?:[.,]\d+)*$")
NUM_INNER = re.compile(r"\d+")
BRACKETS_RE = re.compile(r"(\([^)]+\)|\[[^\]]+\])")

def mask_brackets_with_X(s: str, numeric_ratio=0.6):
    def _f(m):
        inner = m.group(0)[1:-1]
        if not inner.strip(): return m.group(0)
        digits = len(re.findall(r"\d", inner))
        nonalpha = len(re.findall(r"[^가-힣A-Za-z]", inner))
        total = max(1, len(inner))
        ratio = max(digits, nonalpha) / total
        if ratio >= numeric_ratio:
            return m.group(0)[0] + " {X} " + m.group(0)[-1]
        return m.group(0)
    return BRACKETS_RE.sub(_f, s)

# ------ 전국 지역 사전 (시·도/시·군·구 + 주요 상권 일부) ------
LOC_KO = {
    # 17개 시·도
    "서울","부산","대구","인천","광주","대전","울산","세종",
    "경기","강원","충북","충남","전북","전남","경북","경남","제주","제주도",
    # 수도권/경기
    "수원","성남","용인","고양","부천","안양","안산","평택","의정부","김포","파주","남양주","하남","구리","시흥",
    "광명","군포","의왕","오산","과천","이천","여주","양평","동두천","가평","연천","포천",
    # 강원
    "춘천","원주","강릉","동해","속초","삼척","태백","홍천","횡성","영월","평창","정선","철원","화천","양구","인제","고성","양양",
    # 충북
    "청주","충주","제천","보은","옥천","영동","증평","진천","괴산","음성","단양",
    # 충남
    "천안","공주","보령","아산","서산","논산","계룡","당진","금산","부여","서천","청양","홍성","예산","태안",
    # 전북
    "전주","군산","익산","정읍","남원","김제","완주","진안","무주","장수","임실","순창","고창","부안",
    # 전남
    "목포","여수","순천","나주","광양","담양","곡성","구례","고흥","보성","화순","장흥","강진","해남","영암","무안","함평","영광","장성","완도","진도","신안",
    # 경북
    "포항","경주","김천","안동","구미","영주","영천","상주","문경","경산","군위","의성","청송","영양","영덕","청도","고령","성주","칠곡","예천","봉화","울진","울릉",
    # 경남
    "창원","진주","통영","사천","김해","밀양","거제","양산","의령","함안","창녕","고성","남해","하동","산청","함양","거창","합천",
    # 제주
    "제주시","서귀포시",
    # 광역시 하위 구(자주 쓰는 곳 위주)
    "해운대","수영","남구","북구","동구","서구","중구","영도","금정","연제","동래","사하","사상","강서","기장",
    "수성","달서","달성",
    "미추홀","연수","남동","부평","계양","강화","옹진",
    "광산","유성","대덕","울주",
    # 유명 상권/관광지
    "강남","여의도","홍대","명동","동대문","남대문","이태원","서촌","북촌",
    "제주공항","성산","중문","애월","협재"
}

# ------ PERSON 탐지(보수) ------
def detect_person_spans(words, poss):
    person_heads = set()

    def is_nnp(i_word):
        labs = poss[i_word] or set()
        return ("NNP" in labs)
    def has_korean_only(s):
        return bool(re.fullmatch(r"[가-힣]+", s))

    # 영문 풀네임
    for i in range(len(words)-1):
        if not (is_word(words[i]) and is_word(words[i+1])): 
            continue
        a, b = words[i], words[i+1]
        if ENG_CAP.match(a) and ENG_CAP.match(b):
            person_heads.add(i)

    # 한글 이름
    n = len(words); i = 0
    while i < n:
        if not is_word(words[i]): 
            i += 1; continue
        w = words[i]
        if has_korean_only(w):
            # 단일 2~3자 NNP + 성씨
            if 2 <= len(w) <= 3 and w[0] in COMMON_SURNAMES and is_nnp(i):
                person_heads.add(i); i += 1; continue
            # 공백 포함 2~4자, 모두 NNP
            j = i
            merged = w
            while j+1 < n and is_word(words[j+1]) and has_korean_only(words[j+1]) and len(merged) < 4:
                merged += words[j+1]; j += 1
            if 2 <= len(merged) <= 4 and merged[0] in COMMON_SURNAMES:
                ok = True
                for k in range(i, j+1):
                    labs = poss[k] or set()
                    if "NNP" not in labs: ok = False; break
                if ok:
                    person_heads.add(i)
                    i = j + 1
                    continue
        i += 1
    return person_heads

# ------ 각 단어에 대해 Kkma.pos 요약 ------
def pos_labels_per_word(token: str):
    labs = set()
    try:
        for t, p in kkma.pos(token):
            labs.add(p)
    except Exception:
        pass
    return labs

# ------ 메인: 한 제목 → 템플릿 ------
def to_template_surface(title: str,
                        min_noun_len=3,
                        max_mask_ratio=0.55,    # (참고값) 지금은 삭제/복원 안 함
                        max_person_per_title=2,
                        bracket_numeric_ratio=0.6):
    s0 = clean_text(title)
    # 0) URL/EMAIL/HASHTAG 선치환
    s = re.sub(r"https?://\S+|www\.\S+", " {URL} ", s0)
    s = re.sub(r"[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}", " {EMAIL} ", s)
    s = re.sub(r"#\S+", " {HASHTAG} ", s)

    # 1) 분해(표면형 보존)
    parts = split_preserve(s)

    # 2) POS 요약
    labels = [pos_labels_per_word(w) if is_word(w) and w not in {"{URL}","{EMAIL}","{HASHTAG}"} else None
              for w in parts]

    # 3) PERSON 헤드 탐지
    person_heads = detect_person_spans(parts, labels)

    # 4) 치환: PERSON → NUM(조사 보존) → ENG → LOC(조사 보존) → NOUN(조사 보존)
    out = []
    person_used = 0

    for i, w in enumerate(parts):
        if not is_word(w) or w in {"{URL}","{EMAIL}","{HASHTAG}"}:
            out.append(w); continue

        low = w.lower()
        labs = labels[i] or set()

        # 조사 분리(이하 보존 대상에서 사용)
        stem, josa = split_stem_josa(w)

        # 4-1) PERSON (헤드만, 조사 보존 여부는 선택 — 필요 시 아래처럼 붙이기)
        if i in person_heads and person_used < max_person_per_title:
            # out.append("{PERSON}"+josa)  # ← PERSON에도 조사 보존을 원하면 이 줄 사용
            out.append("{PERSON}")         # ← 기본: PERSON은 조사 미보존(요청 시 위로 교체)
            person_used += 1
            continue

        # 4-2) NUM — 조사 보존
        if NUM_FULL.match(stem) or ("SN" in labs) or ("NR" in labs):
            out.append("{NUM}"+josa); continue
        if NUM_INNER.search(stem) and not ENG_WORD.match(stem):
            out.append(NUM_INNER.sub("{NUM}", stem) + josa); continue

        # 4-3) ENG (조사 붙는 케이스 드물지만, 필요 시 보존해도 됨)
        if ENG_WORD.match(stem):
            # out.append("{ENG}"+josa)   # ENG에도 조사 보존 원하면 사용
            out.append("{ENG}"); continue

        # 4-4) LOC — 조사 보존
        if stem in LOC_KO or w in LOC_KO:
            if stem in LOC_KO and josa:
                out.append("{LOC}"+josa); continue
            if w in LOC_KO:
                out.append("{LOC}"); continue

        # 4-5) 상위 키워드/불용어 보호
        if low in keep_tokens or low in STOPWORDS:
            out.append(w); continue

        # 4-6) NOUN — 조사 보존
        if "NNP" in labs:
            out.append("{NOUN}"+josa); continue
        if "NNG" in labs and len(stem) >= min_noun_len:
            out.append("{NOUN}"+josa); continue

        # 4-7) 그 외는 원형 유지
        out.append(w)

    # 5) 과마스킹 완화 — 삭제 없음(그대로 유지)
    #    필요 시 '너무 태그가 많다'고 느껴지면, 이후 단계에서 리랭크만 고려하세요.

    # 6) 조립 & 후처리
    out_str = "".join(out)
    out_str = re.sub(r"(?:\{PERSON\}\s*){2,}", "{PERSON} ", out_str)
    out_str = re.sub(r"(?:\{NUM\}\s*){2,}", "{NUM} ", out_str)
    out_str = re.sub(r"(?:\{NOUN\}\s*){2,}", "{NOUN} ", out_str)
    out_str = re.sub(r"(?:\{LOC\}\s*){2,}", "{LOC} ", out_str)
    out_str = re.sub(r"\s+", " ", out_str).strip()
    out_str = mask_brackets_with_X(out_str, numeric_ratio=bracket_numeric_ratio)
    return out_str

# ---------- 실행 ----------
templates = [to_template_surface(t,
                                 min_noun_len=3,
                                 max_mask_ratio=0.55,
                                 max_person_per_title=2,
                                 bracket_numeric_ratio=0.6)
             for t in tqdm(df_all["title"].astype(str), desc="Templates (surface-safe + LOC/NUM/NOUN josa)")]

df_all["title_template"] = templates

# 요약
template_summary = (
    df_all["title_template"].fillna("").replace("", None).dropna()
      .value_counts().reset_index().rename(columns={"index":"template","title_template":"count"})
)
print(template_summary.head(15))
```

```python
# =========================
# 템플릿 생성 (word-level, 표면형 보존)
# 순서: LOC(조사 보존) → PERSON → NUM(조사 보존) → ENG → NOUN(조사 보존) → 괄호{X}
# =========================
import re, unicodedata
from tqdm import tqdm
from konlpy.tag import Kkma

kkma = Kkma()

# (있으면 사용; 없으면 빈 세트)
keep_tokens = {*(keep_tokens if 'keep_tokens' in globals() else set())}
STOPWORDS  = {*(STOPWORDS  if 'STOPWORDS'  in globals() else set())}

# ------ 유틸: 유니코드 안정화 & 토큰화(표면형 보존) ------
_SURROGATE_RE = re.compile(r"[\ud800-\udfff]")
_ZERO_WIDTH_RE = re.compile(r"[\u200B-\u200F\uFEFF]")
_EMOJI_SUPP_RE = re.compile(r"[\U00010000-\U0010FFFF]")

def clean_text(s: str) -> str:
    s = unicodedata.normalize("NFC", str(s))
    s = _ZERO_WIDTH_RE.sub(" ", s)
    s = _SURROGATE_RE.sub(" ", s)
    s = _EMOJI_SUPP_RE.sub(" ", s)
    return s

# 단어/구두점/공백 분리 (표면형 보존)
WORD_SPLIT = re.compile(r"([^\w가-힣]+)")

def split_preserve(text: str):
    parts = []
    for chunk in WORD_SPLIT.split(text):
        if chunk:
            parts.append(chunk)
    return parts

def is_word(token: str) -> bool:
    return bool(re.search(r"[A-Za-z0-9가-힣]", token))

# ------ 조사 사전(긴 것 우선) & 분리 ------
JOSA_SUFFIXES = [
    "으로써","으로서","으로부터","로부터",
    "이었습니다","였습니다","이었다","였다","이었다가","였다가",
    "이라고는","라고는","이라고","라고",
    "에게서","에게만","에게도","에게","께서","께",
    "에서는","에서","에는만","에는도","에는","에는도","에는",
    "으로는","로는","으로","로",
    "이라면","라면","이라서","라서","이지만","지만","이거나","거나",
    "이며서","면서","이며","며","이고","고",
    "과의","와의","과는","와는","과도","와도","과만","와만","과","와","랑은","랑도","랑만","랑",
    "은","는","이","가","을","를","도","만","뿐","까지","부터","마저","조차","이나","나","의"
]
JOSA_SUFFIXES = sorted({j for j in JOSA_SUFFIXES if re.fullmatch(r"[가-힣]+", j)}, key=len, reverse=True)

def split_stem_josa(word: str):
    """어절 끝의 조사 1개를 떼어 (stem, josa)로 반환. 조사 없으면 (word, "")"""
    for suf in JOSA_SUFFIXES:
        if word.endswith(suf) and len(word) > len(suf):
            return word[:-len(suf)], suf
    return word, ""

# ------ 패턴 & 사전 ------
COMMON_SURNAMES = {
    "김","이","박","최","정","강","조","윤","장","임","한","오","서","신","권","황","안","송","전","홍","유","고","문",
    "양","손","배","백","허","남","심","노","하","곽","성","차","주","우","구","민","류","나","진","지","엄","채","원",
    "천","방","공","현","함","변","염","여","추","도","소","석","선","설","마","길","연","위","표","명","기","반","라",
    "왕","금","옥","육","인","맹","제","모","남궁","탁","국","어"
}
ENG_WORD  = re.compile(r"^[A-Za-z]{2,}$")
ENG_CAP   = re.compile(r"^[A-Z][a-z]{1,20}$")
NUM_FULL  = re.compile(r"^\d+(?:[.,]\d+)*$")
NUM_INNER = re.compile(r"\d+")
BRACKETS_RE = re.compile(r"(\([^)]+\)|\[[^\]]+\])")

def mask_brackets_with_X(s: str, numeric_ratio=0.6):
    def _f(m):
        inner = m.group(0)[1:-1]
        if not inner.strip(): return m.group(0)
        digits = len(re.findall(r"\d", inner))
        nonalpha = len(re.findall(r"[^가-힣A-Za-z]", inner))
        total = max(1, len(inner))
        ratio = max(digits, nonalpha) / total
        if ratio >= numeric_ratio:
            return m.group(0)[0] + " {X} " + m.group(0)[-1]
        return m.group(0)
    return BRACKETS_RE.sub(_f, s)

# ------ 한국 전국 지역 사전 (시·도/시·군·구 + 주요 상권 일부) ------
LOC_KO = {
    "서울","부산","대구","인천","광주","대전","울산","세종",
    "경기","강원","충북","충남","전북","전남","경북","경남","제주","제주도",
    "수원","성남","용인","고양","부천","안양","안산","평택","의정부","김포","파주","남양주","하남","구리","시흥",
    "광명","군포","의왕","오산","과천","이천","여주","양평","동두천","가평","연천","포천",
    "춘천","원주","강릉","동해","속초","삼척","태백","홍천","횡성","영월","평창","정선","철원","화천","양구","인제","고성","양양",
    "청주","충주","제천","보은","옥천","영동","증평","진천","괴산","음성","단양",
    "천안","공주","보령","아산","서산","논산","계룡","당진","금산","부여","서천","청양","홍성","예산","태안",
    "전주","군산","익산","정읍","남원","김제","완주","진안","무주","장수","임실","순창","고창","부안",
    "목포","여수","순천","나주","광양","담양","곡성","구례","고흥","보성","화순","장흥","강진","해남","영암","무안","함평","영광","장성","완도","진도","신안",
    "포항","경주","김천","안동","구미","영주","영천","상주","문경","경산","군위","의성","청송","영양","영덕","청도","고령","성주","칠곡","예천","봉화","울진","울릉",
    "창원","진주","통영","사천","김해","밀양","거제","양산","의령","함안","창녕","고성","남해","하동","산청","함양","거창","합천",
    "제주시","서귀포시",
    
    "해운대","수영","남구","북구","동구","서구","중구","영도","금정","연제","동래","사하","사상","강서","기장",
    "수성","달서","달성",
    "미추홀","연수","남동","부평","계양","강화","옹진",
    "광산","유성","대덕","울주",
    "강남","여의도","홍대","명동","동대문","남대문","이태원","서촌","북촌",
    "제주공항","성산","중문","애월","협재"
}

# ------ 각 단어에 대해 Kkma.pos 요약 ------
def pos_labels_per_word(token: str):
    labs = set()
    try:
        for t, p in kkma.pos(token):
            labs.add(p)
    except Exception:
        pass
    return labs

# ------ PERSON 탐지(보수) — LOC 제외 반영 ------
def detect_person_spans(words, poss, loc_set=None):
    """
    loc_set: LOC_KO 집합을 넘기면, 해당 단어/병합어가 LOC일 경우 PERSON 후보에서 제외
    """
    if loc_set is None:
        loc_set = set()
    person_heads = set()

    def is_nnp(i_word):
        labs = poss[i_word] or set()
        return ("NNP" in labs)

    def has_korean_only(s):
        return bool(re.fullmatch(r"[가-힣]+", s))

    # 영문 풀네임 (LOC 제외)
    for i in range(len(words)-1):
        if not (is_word(words[i]) and is_word(words[i+1])): 
            continue
        a, b = words[i], words[i+1]
        if (a in loc_set) or (b in loc_set):
            continue
        if ENG_CAP.match(a) and ENG_CAP.match(b):
            person_heads.add(i)

    # 한글 이름 (LOC 제외)
    n = len(words); i = 0
    while i < n:
        if not is_word(words[i]): 
            i += 1; continue
        w = words[i]
        if w in loc_set:           # 지역 단어는 스킵
            i += 1; continue
        if has_korean_only(w):
            # 단일 2~3자 NNP + 성씨
            if 2 <= len(w) <= 3 and w[0] in COMMON_SURNAMES and is_nnp(i) and (w not in loc_set):
                person_heads.add(i); i += 1; continue
            # 공백 포함 2~4자, 모두 NNP
            j = i
            merged = w
            while j+1 < n and is_word(words[j+1]) and has_korean_only(words[j+1]) and len(merged) < 4:
                merged += words[j+1]; j += 1
            if (merged not in loc_set) and 2 <= len(merged) <= 4 and merged[0] in COMMON_SURNAMES:
                ok = True
                for k in range(i, j+1):
                    labs = poss[k] or set()
                    if "NNP" not in labs: ok = False; break
                if ok:
                    person_heads.add(i)
                    i = j + 1
                    continue
        i += 1
    return person_heads

# ------ 메인: 한 제목 → 템플릿 ------
def to_template_surface(title: str,
                        min_noun_len=3,
                        max_mask_ratio=0.55,    # 현재는 삭제/복원 없음
                        max_person_per_title=2,
                        bracket_numeric_ratio=0.6):
    s0 = clean_text(title)
    # 0) URL/EMAIL/HASHTAG 선치환
    s = re.sub(r"https?://\S+|www\.\S+", " {URL} ", s0)
    s = re.sub(r"[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}", " {EMAIL} ", s)
    s = re.sub(r"#\S+", " {HASHTAG} ", s)

    # 1) 분해(표면형 보존)
    parts = split_preserve(s)

    # 2) POS 요약
    labels = [pos_labels_per_word(w) if is_word(w) and w not in {"{URL}","{EMAIL}","{HASHTAG}"} else None
              for w in parts]

    # 3) PERSON 헤드 인덱스 (LOC 제외 반영)
    person_heads = detect_person_spans(parts, labels, loc_set=LOC_KO)

    # 4) 치환: LOC(조사 보존) → PERSON → NUM(조사 보존) → ENG → NOUN(조사 보존)
    out = []
    person_used = 0

    for i, w in enumerate(parts):
        if not is_word(w) or w in {"{URL}","{EMAIL}","{HASHTAG}"}:
            out.append(w); continue

        low = w.lower()
        labs = labels[i] or set()

        # 조사 분리 (조사 보존용)
        stem, josa = split_stem_josa(w)

        # 4-1) LOC — 먼저 처리 (조사 보존)
        if stem in LOC_KO or w in LOC_KO:
            if stem in LOC_KO and josa:
                out.append("{LOC}"+josa); continue
            if w in LOC_KO:
                out.append("{LOC}"); continue

        # 4-2) PERSON — LOC 태깅 뒤 수행
        if i in person_heads and person_used < max_person_per_title:
            # PERSON에도 조사 보존을 원하면 아래 주석 해제
            # out.append("{PERSON}"+josa); person_used += 1; continue
            out.append("{PERSON}"); person_used += 1; continue

        # 4-3) NUM — 조사 보존
        if NUM_FULL.match(stem) or ("SN" in labs) or ("NR" in labs):
            out.append("{NUM}"+josa); continue
        if NUM_INNER.search(stem) and not ENG_WORD.match(stem):
            out.append(NUM_INNER.sub("{NUM}", stem) + josa); continue

        # 4-4) ENG (필요 시 조사 보존으로 바꿀 수 있음)
        if ENG_WORD.match(stem):
            # out.append("{ENG}"+josa); continue
            out.append("{ENG}"); continue

        # 4-5) 상위 키워드/불용어 보호
        if low in keep_tokens or low in STOPWORDS:
            out.append(w); continue

        # 4-6) NOUN — 조사 보존
        if "NNP" in labs:
            out.append("{NOUN}"+josa); continue
        if "NNG" in labs and len(stem) >= min_noun_len:
            out.append("{NOUN}"+josa); continue

        # 4-7) 그 외(동사/형용사/부사/감탄사/조사 단독 등)는 원형 유지
        out.append(w)

    # 5) 과마스킹 완화 — 삭제 없음(표면형 보존)
    out_str = "".join(out)
    out_str = re.sub(r"(?:\{PERSON\}\s*){2,}", "{PERSON} ", out_str)
    out_str = re.sub(r"(?:\{NUM\}\s*){2,}", "{NUM} ", out_str)
    out_str = re.sub(r"(?:\{NOUN\}\s*){2,}", "{NOUN} ", out_str)
    out_str = re.sub(r"(?:\{LOC\}\s*){2,}", "{LOC} ", out_str)
    out_str = re.sub(r"\s+", " ", out_str).strip()

    # 6) 괄호 {X} 처리 (숫자/기호 위주일 때만)
    out_str = mask_brackets_with_X(out_str, numeric_ratio=bracket_numeric_ratio)
    return out_str

# ---------- 실행 ----------
templates = [to_template_surface(
                t,
                min_noun_len=3,
                max_mask_ratio=0.55,
                max_person_per_title=2,
                bracket_numeric_ratio=0.6
            ) for t in tqdm(df_all["title"].astype(str), desc="Templates (LOC → PERSON → NUM → ENG → NOUN)")]

df_all["title_template"] = templates

# 요약 확인
template_summary = (
    df_all["title_template"].fillna("").replace("", None).dropna()
    .value_counts().reset_index().rename(columns={"index":"template","title_template":"count"})
)
print(template_summary.head(15))
```
