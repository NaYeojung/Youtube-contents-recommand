```python
import re, numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer

# 0) ì…ë ¥ ê°€ì •
# df_all: ìœ„ ë‹¨ê³„ì—ì„œ ë§Œë“  ê²°í•© DF (í•„ìˆ˜: 'title')
# í•„ìš”í•˜ë©´ ì¤‘ë³µ/ê²°ì¸¡ ì œê±°
df_all = df_all.copy()
df_all = df_all.dropna(subset=["title"])
df_all = df_all.drop_duplicates(subset=["title"]).reset_index(drop=True)

# 1) ê°„ë‹¨ í† í¬ë‚˜ì´ì € + ë¶ˆìš©ì–´
def tokenize_ko(text: str):
    text = re.sub(r"[^0-9ê°€-í£A-Za-z]+", " ", str(text))
    toks = text.lower().split()
    return [t for t in toks if len(t) > 1]
STOPWORDS = set(["ì˜ìƒ","ì±„ë„","ìœ íŠœë¸Œ"])  # í•„ìš”ì‹œ ì¶”ê°€
def analyzer(doc): return [t for t in tokenize_ko(doc) if t not in STOPWORDS]

# 2) TF-IDF ì í•©
vectorizer = TfidfVectorizer(
    analyzer=analyzer,
    ngram_range=(1,2),  # ë‹¨ì–´+ë°”ì´ê·¸ë¨
    min_df=2            # 2ê°œ ì´ìƒ ë¬¸ì„œì—ì„œ ë“±ì¥í•œ í† í°ë§Œ
)
X = vectorizer.fit_transform(df_all["title"].astype(str))
vocab = np.array(vectorizer.get_feature_names_out())

# 3) ì „ì²´ ìƒìœ„ í‚¤ì›Œë“œ(top-K)
top_k = 30
scores = np.asarray(X.sum(axis=0)).ravel()
order = scores.argsort()[::-1][:top_k]
top_keywords = list(zip(vocab[order], scores[order]))

print("[ì „ì²´ ìƒìœ„ í‚¤ì›Œë“œ]")
for k, s in top_keywords:
    print(f"{k}\t{s:.3f}")
```

```python
import re
import random
import itertools
import math
import pandas as pd
from collections import Counter, defaultdict
from konlpy.tag import Okt

okt = Okt()

############################################
# 1) í‚¤ì›Œë“œ í’€ ë§Œë“¤ê¸° (KeyBERT âˆª TF-IDF)
############################################

def normalize_text(s: str) -> str:
    s = (s or "").strip()
    s = re.sub(r'\s+', ' ', s)
    return s

def is_korean(word: str) -> bool:
    return bool(re.search(r'[ê°€-í£]', word))

def pick_nouns(phrase: str):
    # í‚¤ì›Œë“œ/êµ¬ì—ì„œ í•µì‹¬ ëª…ì‚¬ë§Œ ì¶”ì¶œ (ë„ˆë¬´ ì§§ì€ ê±´ ì œì™¸)
    return [n for n in okt.nouns(phrase) if len(n) >= 2]

def build_keyword_pool(enriched: pd.DataFrame,
                       w_keybert: float = 1.0,
                       w_tfidf: float = 0.5,
                       top_per_title: int = 5,
                       min_len: int = 2):
    """
    corpus ì „ì²´ì—ì„œ ì‚¬ìš©í•  'í‚¤ì›Œë“œ í’€'ê³¼ ë¬¸ì„œë³„(ì œëª©ë³„) í‚¤ì›Œë“œ í’€ì„ ë§Œë“ ë‹¤.
    - KeyBERT(ì˜ë¯¸) ìš°ì„ , TF-IDF(í˜•íƒœ) ë³´ì¡°
    """
    global_pool = defaultdict(float)
    per_title_pool = []

    for _, row in enriched.iterrows():
        kb = row.get("keybert_keywords") or []
        tf = row.get("tfidf_keywords") or []

        # 1) KeyBERT: (êµ¬, ì ìˆ˜)
        kb_pairs = [(normalize_text(k), float(s)) for k, s in kb][:top_per_title]
        # 2) TF-IDF: (ì¡°ê°, ì ìˆ˜) -> ëª…ì‚¬ë§Œ ì¶”ì¶œ
        tf_pairs = []
        for k, s in (tf[:top_per_title] if isinstance(tf, list) else []):
            k = normalize_text(k)
            if not k: 
                continue
            # í•œê¸€ ì•„ë‹ˆë©´ íŒ¨ìŠ¤(ë¶ˆìš© ASCII ì¡°ê° ì œê±°)
            if not is_korean(k):
                continue
            for n in pick_nouns(k):
                tf_pairs.append((n, float(s)))

        # per-title pool ë§Œë“¤ê¸° (ë™ì¼ í‚¤ ì¤‘ë³µ í•©ì‚°)
        local_scores = defaultdict(float)
        for k, s in kb_pairs:
            # êµ¬ë¥¼ ëª…ì‚¬ë¡œ ìª¼ê°  ë²„ì „ + ì›êµ¬ ëª¨ë‘ ë°˜ì˜
            nouns = pick_nouns(k)
            if nouns:
                for n in nouns:
                    local_scores[n] += w_keybert * s
            else:
                if is_korean(k) and len(k) >= min_len:
                    local_scores[k] += w_keybert * s
        for k, s in tf_pairs:
            local_scores[k] += w_tfidf * s

        # ì •ë¦¬
        local_sorted = sorted(local_scores.items(), key=lambda x: x[1], reverse=True)
        per_title_pool.append(local_sorted)
        for k, s in local_sorted:
            global_pool[k] += s

    # ê¸€ë¡œë²Œ í’€ ì •ë ¬
    global_sorted = sorted(global_pool.items(), key=lambda x: x[1], reverse=True)
    return global_sorted, per_title_pool

############################################
# 2) ë¼ë²¨/ë¸Œë¼ì¼“ {X} í’€ ë§Œë“¤ê¸°
############################################

def extract_bracket_labels(titles):
    """
    ì œëª©ì—ì„œ [ë¼ë²¨], (ë¶€ê°€ì •ë³´) ë“± ë¸Œë¼ì¼“ ì½˜í…ì¸  ì¶”ì¶œ â†’ {X} í›„ë³´ í’€
    """
    br_pat = re.compile(r'(\[[^\]]+\]|\([^)]+\)|\{[^}]+\})')
    counts = Counter()
    items = []
    for t in titles:
        t = t or ""
        for m in br_pat.findall(t):
            val = m.strip()
            # ë„ˆë¬´ ê¸¸ê±°ë‚˜ ì¡ìŒ ì‹¬í•˜ë©´ ì œì™¸
            if 2 <= len(val) <= 40:
                items.append(val)
                counts[val] += 1
    # ìƒìœ„ ë¼ë²¨
    top_labels = [x for x,_ in counts.most_common(100)]
    return top_labels, counts

############################################
# 3) í…œí”Œë¦¿ ì„ íƒ
############################################

def select_top_templates(enriched: pd.DataFrame, top_n=20):
    vc = (enriched['title_template'].dropna()
          .map(lambda s: s.strip())
          .value_counts())
    return vc.index[:top_n].tolist()

############################################
# 4) í•œêµ­ì–´ ì¡°ì‚¬ ë³´ì • (ê°„ë‹¨ ê·œì¹™)
############################################

def has_final_consonant(k: str) -> bool:
    # í•œê¸€ ë§ˆì§€ë§‰ ê¸€ìì˜ ì¢…ì„± ìœ ë¬´ë¡œ ì¡°ì‚¬ ì„ íƒ
    if not k: return False
    ch = k[-1]
    if not ('ê°€' <= ch <= 'í£'):
        return False
    code = ord(ch) - ord('ê°€')
    jong = code % 28
    return jong != 0

def attach_josa(word: str, pair=("ì€","ëŠ”")) -> str:
    # pair ì˜ˆ: ("ì€","ëŠ”"), ("ì´","ê°€"), ("ì„","ë¥¼"), ("ê³¼","ì™€")
    return word + (pair[0] if has_final_consonant(word) else pair[1])

############################################
# 5) ìŠ¬ë¡¯ ì±„ìš°ê¸° + ìƒì„±
############################################

def fill_template(template: str,
                  noun_pool: list,
                  label_pool: list,
                  num_pool: list = None,
                  rng=None):
    """
    templateì˜ {NOUN}, {X}, {NUM}, {ENG} ìŠ¬ë¡¯ì„ ì±„ì›Œ ì¶”ì²œ ì œëª© ìƒì„±.
    noun_pool, label_poolëŠ” (í‚¤, ì ìˆ˜) ë¦¬ìŠ¤íŠ¸.
    """
    if rng is None:
        rng = random.Random()

    t = template

    # í›„ë³´ ë¦¬ìŠ¤íŠ¸ ë§Œë“¤ê¸°
    nouns = [k for k,_ in noun_pool]
    labels = label_pool or []

    # ê°„ë‹¨í•œ ìˆ«ì í›„ë³´ (í•„ìš”ì‹œ ì™¸ë¶€ì—ì„œ ì „ë‹¬)
    if num_pool is None:
        num_pool = ["3", "5", "7", "10", "2025"]

    # {X} ì±„ìš°ê¸°: ê°€ì¥ ë¹ˆë„ ë†’ì€ ë¼ë²¨ë“¤
    while "{X}" in t and labels:
        lab = labels[min(rng.randrange(len(labels)), 5)]  # ë„ˆë¬´ ìƒìœ„ë§Œ ë°˜ë³µë˜ì§€ ì•Šê²Œ
        t = t.replace("{X}", lab, 1)

    # {NUM} ì±„ìš°ê¸°
    while "{NUM}" in t and num_pool:
        n = rng.choice(num_pool)
        t = t.replace("{NUM}", n, 1)

    # {ENG}ëŠ” ë³´í†µ ë¼ë²¨ë¡œ ëŒ€ì²´í•˜ê±°ë‚˜ ì œê±°
    t = t.replace("{ENG}", "")

    # {NOUN} ì±„ìš°ê¸°: ìƒìœ„ì—ì„œ ëœë¤ ìƒ˜í”Œ(ì¤‘ë³µ ìµœì†Œí™”)
    used = set()
    def pick_noun():
        for k,_ in noun_pool:
            if k not in used:
                used.add(k); return k
        # ë‹¤ ì“°ë©´ ì¬ì‚¬ìš© í—ˆìš©
        return noun_pool[0][0] if noun_pool else "ì½˜í…ì¸ "

    while "{NOUN}" in t:
        t = t.replace("{NOUN}", pick_noun(), 1)

    # ì¡°ì‚¬ ë³´ì •: ê°„ë‹¨ íŒ¨í„´ë§Œ ì²˜ë¦¬
    # ...{NOUN}ì€/ëŠ”, {NOUN}ì´/ê°€, {NOUN}ì„/ë¥¼ ë“± íŒ¨í„´ êµì •ì´ í•„ìš”í•˜ë©´
    t = re.sub(r'(\S+)\s*ì€/ëŠ”', lambda m: attach_josa(m.group(1), ("ì€","ëŠ”")), t)
    t = re.sub(r'(\S+)\s*ì´/ê°€', lambda m: attach_josa(m.group(1), ("ì´","ê°€")), t)
    t = re.sub(r'(\S+)\s*ì„/ë¥¼', lambda m: attach_josa(m.group(1), ("ì„","ë¥¼")), t)
    t = re.sub(r'(\S+)\s*ê³¼/ì™€', lambda m: attach_josa(m.group(1), ("ê³¼","ì™€")), t)

    # ê³µë°±/êµ¬ë‘ì  ì •ë¦¬
    t = re.sub(r'\s+', ' ', t).strip()
    t = re.sub(r'\s+([),\].!?])', r'\1', t)
    t = re.sub(r'([(\[])\s+', r'\1', t)
    t = re.sub(r'\s+([:;])\s*', r'\1 ', t)
    # ì—°ì† ë§ˆì¹¨í‘œ/ë”°ì˜´í‘œ ì •ë¦¬
    t = re.sub(r'\.{3,}', '...', t)
    t = re.sub(r'["â€œâ€]{2,}', '"', t)

    # ê´„í˜¸ ë¶ˆê· í˜• ê°„ë‹¨ ë³´ì •
    if t.count('(') != t.count(')'):
        t = t.replace('(', '').replace(')', '')
    if t.count('[') != t.count(']'):
        t = t.replace('[', '').replace(']', '')

    return normalize_text(t)

############################################
# 6) í›„ë³´ ìƒì„± + ì ìˆ˜í™” + ë‹¤ì–‘ì„±(MMR ìœ ì‚¬)
############################################

def title_length_score(title, target_range=(18, 35)):
    L = len(title)
    a,b = target_range
    if a <= L <= b: return 1.0
    # ê±°ë¦¬ ê¸°ë°˜ íŒ¨ë„í‹°
    d = min(abs(L-a), abs(L-b))
    return 1.0 / (1.0 + 0.08*d)

def keyword_coverage_score(title, noun_pool, top_k=10):
    top = set(k for k,_ in noun_pool[:top_k])
    hit = sum(1 for k in top if k in title)
    return hit / max(1, len(top))

def diversity_penalty(new_title, chosen, beta=0.7):
    # ê°„ë‹¨ Jaccard ê¸°ë°˜ íŒ¨ë„í‹°
    def shingles(s):
        return set([s[i:i+3] for i in range(max(1, len(s)-2))])
    A = shingles(new_title)
    if not chosen: return 1.0
    sims = []
    for t in chosen:
        B = shingles(t)
        inter = len(A & B)
        union = len(A | B) or 1
        sims.append(inter/union)
    max_sim = max(sims)
    return (1.0 - beta*max_sim)

def score_title(title, noun_pool):
    return 0.5*keyword_coverage_score(title, noun_pool) + 0.5*title_length_score(title)

def generate_recommended_titles(enriched: pd.DataFrame,
                                top_templates: list = None,
                                num_candidates_per_template: int = 8,
                                total_return: int = 10,
                                seed: int = 42):
    rng = random.Random(seed)

    # 1) ì „ì—­ í‚¤ì›Œë“œ í’€ + ë¼ë²¨í’€
    global_pool, per_title_pool = build_keyword_pool(enriched)
    labels, _ = extract_bracket_labels(enriched['title'].tolist())

    # 2) í…œí”Œë¦¿ ì„ íƒ
    if top_templates is None:
        top_templates = select_top_templates(enriched, top_n=20)

    # 3) í…œí”Œë¦¿ë³„ í›„ë³´ ìƒì„±
    candidates = []
    for tpl in top_templates:
        noun_pool = global_pool[:30]  # ìƒìœ„ 30ê°œë¡œ ìŠ¬ë¡¯ ì±„ìš°ê¸°
        for _ in range(num_candidates_per_template):
            cand = fill_template(tpl, noun_pool=noun_pool, label_pool=labels, rng=rng)
            sc = score_title(cand, noun_pool)
            candidates.append((cand, tpl, sc))

    # 4) ì ìˆ˜ìˆœ ì •ë ¬ í›„ MMRì‹ ë‹¤ì–‘ì„± í•„í„°
    candidates.sort(key=lambda x: x[2], reverse=True)
    chosen = []
    used = set()
    for cand, tpl, sc in candidates:
        if cand in used:
            continue
        pen = diversity_penalty(cand, [t[0] for t in chosen], beta=0.7)
        final = sc * pen
        chosen.append((cand, tpl, sc, final))
        used.add(cand)
        if len(chosen) >= total_return:
            break

    out = pd.DataFrame(chosen, columns=["title_reco","source_template","base_score","final_score"])
    return out.sort_values("final_score", ascending=False).reset_index(drop=True)
```

```python
reco_df = generate_recommended_titles(enriched,
                                      top_templates=None,      # Noneì´ë©´ ìƒìœ„ í…œí”Œë¦¿ ìë™ ì„ íƒ
                                      num_candidates_per_template=8,
                                      total_return=12,
                                      seed=7)

print(reco_df)
```
