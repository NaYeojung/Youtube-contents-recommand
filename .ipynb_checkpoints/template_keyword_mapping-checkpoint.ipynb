## 키워드 점수 합산

```python
import pandas as pd

df_k=pd.read_csv('data/channel(@KoreanCryingGuy)_title_keyword.csv', encoding='utf-8')
df_k=df_k.iloc[:,[0,19,20]]
df_k
```

```python
import ast

# 문자열을 실제 리스트/튜플로 변환
df_k['tfidf_keywords'] = df_k['tfidf_keywords'].apply(ast.literal_eval)
df_k['keybert_keywords'] = df_k['keybert_keywords'].apply(ast.literal_eval)
```

```python
from konlpy.tag import Okt
import json

with open("title_template.json", "r", encoding="utf-8") as f:
    gpt_templates = json.load(f)

okt = Okt()

target_slots = {"PERSON", "TOPIC", "EVENT", "LOC"}

def extract_nouns(kw):
    return okt.nouns(kw)

for entry in gpt_templates:
    slots = entry.get("slots", {})
    for slot_name, slot_value in slots.items():
        if slot_name in target_slots:
            # slot_value가 문자열이면 명사형으로 변경
            if isinstance(slot_value, str):
                nouns = extract_nouns(slot_value)
                # 명사 여러 개일 경우 공백으로 연결
                entry["slots"][slot_name] = " ".join(nouns)
            # 리스트인 경우 각 항목을 처리
            elif isinstance(slot_value, list):
                new_values = []
                for v in slot_value:
                    new_values.extend(extract_nouns(v))
                entry["slots"][slot_name] = new_values
```

```python
# 각 제목별 GPT 키워드 매핑
gpt_keywords_dict = {item['original_title']: item.get('keywords', []) for item in gpt_templates}
gpt_keywords_dict
```

```python
df_k['gpt_keywords'] = df_k['title'].apply(lambda t: gpt_keywords_dict.get(t, []))
```

```python
def combine_all_keywords(tfidf_list, keybert_list, gpt_list, top_n=5):
    combined = {}

    # TF-IDF
    for item in tfidf_list:
        if isinstance(item, (list, tuple)) and len(item) == 2:
            word, score = item
        else:
            word, score = item, 1.0
        combined[word] = combined.get(word, 0) + score

    # KeyBERT
    for item in keybert_list:
        if isinstance(item, (list, tuple)) and len(item) == 2:
            word, score = item
        else:
            word, score = item, 1.0
        combined[word] = combined.get(word, 0) + score

    # GPT 키워드
    for word in gpt_list:
        combined[word] = combined.get(word, 0) + 1.0 / len(gpt_list)

    # 상위 top_n 선택
    sorted_keywords = sorted(combined.items(), key=lambda x: x[1], reverse=True)
    return [word for word, _ in sorted_keywords[:top_n]]
```

```python
df_k['final_keywords'] = df_k.apply(
    lambda row: combine_all_keywords(
        row['tfidf_keywords'],
        row['keybert_keywords'],
        gpt_keywords_dict.get(row['title'], []),
        top_n=5
    ),
    axis=1
)
df_k
```

```python
from collections import Counter

all_keywords = []

for kws in df_k['final_keywords']:
    if isinstance(kws, list):
        all_keywords.extend(kws)

# 등장 횟수 계산
keyword_counts = Counter(all_keywords)

df_keyword_counts = pd.DataFrame(keyword_counts.items(), columns=['keyword', 'count'])

# 내림차순 정렬
df_keyword_counts = df_keyword_counts.sort_values(by='count', ascending=False).reset_index(drop=True)
df_keyword_counts = df_keyword_counts[df_keyword_counts['count'] >= 2]
df_keyword_counts
```

```python
df_keyword_counts.to_csv("title_keyword_counts.csv", index=False, encoding='utf-8-sig')
```

# 템플릿 합산

## count 기반 상위 템플릿 추출

```python
import pandas as pd
import json

# JSON 파일 불러오기
with open("title_template.json", "r", encoding="utf-8") as f:
    templates_json = json.load(f)

# DataFrame으로 변환
df_templates = pd.DataFrame(templates_json)
df_templates.head()
```

```python
template_counts = df_templates['template'].value_counts()
top_templates = template_counts[template_counts >= 4]
top_templates
```

```python
top_templates_list = top_templates.index.tolist()

# 각 템플릿별 실제 제목들 확인
for template in top_templates_list:
    matched_titles = df_templates[df_templates['template'] == template]['original_title'].tolist()
    print(f"Template: {template}")
    print("Sample Titles:")
    for t in matched_titles:
        print("  -", t)
    print()
```

## 템플릿에 키워드 매핑

### 슬롯별 키워드 랜덤

```python
import random
import pandas as pd

recommendations = []

n_candidates_per_template = 5  # 템플릿당 생성할 후보 제목 수

for template in top_templates_list:
    group = df_templates[df_templates['template'] == template]

    # 모든 슬롯 이름 수집 (행별로 존재하는 키를 모두 포함)
    slot_names = set()
    for d in group['slots']:
        slot_names.update(d.keys())

    # 슬롯별 키워드 수집 (중복 제거)
    slot_keywords = {}
    for slot in slot_names:
        slot_values = [d[slot] for d in group['slots'] if slot in d and d[slot] is not None]
        slot_keywords[slot] = list(set(slot_values))

    # 후보 제목 생성
    for _ in range(n_candidates_per_template):
        chosen_keywords = {}
        for slot, values in slot_keywords.items():
            if values:
                chosen_keywords[slot] = str(random.choice(values))
            else:
                chosen_keywords[slot] = ""  # 값이 없으면 빈 문자열 처리

        # 템플릿에 키워드 채워서 제목 생성
        title_candidate = template
        for slot, kw in chosen_keywords.items():
            title_candidate = title_candidate.replace(f"{{{slot}}}", kw)

        recommendations.append({
            "template": template,
            "title_candidate": title_candidate,
            "slot_mapping": chosen_keywords
        })

df_recommendations = pd.DataFrame(recommendations)
df_recommendations
```

### 슬롯별 키워드 점수 기반

```python
import pandas as pd
from collections import defaultdict

records = []

for idx, row in df_k.iterrows():
    title = row['title']
    tfidf_list = row['tfidf_keywords']  # [('word', score), ...]
    keybert_list = row['keybert_keywords']  # [('word', score), ...]
    gpt_list = row['gpt_keywords'] if 'gpt_keywords' in row else []

    combined_scores = defaultdict(float)

    # TF-IDF 점수 추가
    for item in tfidf_list:
        if isinstance(item, (list, tuple)) and len(item) == 2:
            word, score = item
        else:
            word, score = item, 1.0
        combined_scores[word] += score

    # KeyBERT 점수 추가
    for item in keybert_list:
        if isinstance(item, (list, tuple)) and len(item) == 2:
            word, score = item
        else:
            word, score = item, 1.0
        combined_scores[word] += score

    # GPT 키워드 점수 추가
    for kw in gpt_list:
        combined_scores[kw] += 1.0 / len(gpt_list) if len(gpt_list) > 0 else 0

    # records에 저장
    for word, score in combined_scores.items():
        records.append({
            'title': title,
            'keyword': word,
            'score': score
        })

# 데이터프레임 생성
df_keyword_scores = pd.DataFrame(records)

# 단어별 총점 계산
df_keyword_scores_summary = df_keyword_scores.groupby('keyword')['score'].sum().reset_index()
df_keyword_scores_summary = df_keyword_scores_summary.sort_values(by='score', ascending=False).reset_index(drop=True)

records = []

# df_templates의 각 행에서 슬롯별 단어를 기록
for _, row in df_templates.iterrows():
    slots = row['slots']  # 예: {'SERIES': 'KIA 타이거즈', 'TOPIC': '응원가 제안서'}
    for slot, word in slots.items():
        records.append({"keyword": word, "slot": slot})

df_slots = pd.DataFrame(records)
df_slots = df_slots.drop_duplicates()

# 기존 df_keyword_scores_summary와 합치기
df_keyword_scores_summary = df_keyword_scores_summary.merge(df_slots, on="keyword", how="left")

# # slot 컬럼 정리
# df_keyword_scores_summary['slot'] = (
#     df_keyword_scores_summary['slot']
#     .combine_first(df_keyword_scores_summary['slot_x'])
#     .combine_first(df_keyword_scores_summary['slot_y'])
# )

# 불필요한 중복 컬럼 제거
df_keyword_scores_summary = df_keyword_scores_summary.drop(columns=['slot_x', 'slot_y'], errors='ignore')

# 중복 제거 후 결과 확인
df_keyword_scores_summary = df_keyword_scores_summary.drop_duplicates(subset=['keyword', 'slot'])
df_keyword_scores_summary.reset_index(drop=True, inplace=True)

df_keyword_scores_summary
```

```python
from itertools import product
import pandas as pd

recommendations = []
n_top_keywords = 5  # 슬롯별 상위 키워드 후보 수

for template in top_templates_list:
    group = df_templates[df_templates['template'] == template]

    # 모든 슬롯 이름 수집
    slot_names = set()
    for d in group['slots']:
        slot_names.update(d.keys())

    # 슬롯별 상위 키워드와 점수 수집
    slot_keywords = {}
    slot_scores = {}
    for slot in slot_names:
        slot_values = [d[slot] for d in group['slots'] if slot in d and d[slot] is not None]
        slot_values = list(set(slot_values))
        
        ranked_keywords = df_keyword_scores_summary[df_keyword_scores_summary['keyword'].isin(slot_values)]
        ranked_keywords = ranked_keywords.sort_values(by='score', ascending=False).head(n_top_keywords)

        keywords = ranked_keywords['keyword'].tolist() if not ranked_keywords.empty else [""]
        scores = ranked_keywords['score'].tolist() if not ranked_keywords.empty else [0.0]

        # 중복 제거 + 최대 개수 제한한
        keywords = list(dict.fromkeys(keywords))[:n_top_keywords]
        scores = scores[:len(keywords)]

        slot_keywords[slot] = keywords
        slot_scores[slot] = scores

    # 빈 슬롯 자동 보충 (상위 n개 + 중복 제거)
    for slot in slot_names:
        if not slot_keywords[slot] or all(k == "" for k in slot_keywords[slot]):
            fallback_df = df_keyword_scores_summary[df_keyword_scores_summary['slot'] == slot]
            fallback_df = fallback_df.sort_values(by='score', ascending=False)

            if not fallback_df.empty:
                fallback_keywords = fallback_df['keyword'].tolist()
                fallback_keywords = list(dict.fromkeys(fallback_keywords))[:n_top_keywords]
                fallback_scores = fallback_df['score'].tolist()[:len(fallback_keywords)]
            else:
                fallback_keywords = [f"{{{slot}}}"]
                fallback_scores = [0.0]

            slot_keywords[slot] = fallback_keywords
            slot_scores[slot] = fallback_scores

    # 모든 조합 생성 (각 슬롯별 상위 n_top_keywords까지)
    all_combinations = list(product(*slot_keywords.values()))

    for combo in all_combinations:
        title_candidate = template
        chosen_scores = []

        for slot, kw in zip(slot_names, combo):
            # 슬롯별 맞춤 처리
            if slot == "NUM":
                kw = "00"  # 숫자는 모두 00으로 대체
            elif slot == "SERIES":
                kw = f"[{kw.strip()}]" if kw and not kw.startswith("[") else kw

            # 태그 치환
            title_candidate = title_candidate.replace(f"{{{slot}}}", kw)

            # 점수 가져오기
            if kw in slot_keywords[slot]:
                idx = slot_keywords[slot].index(kw)
                chosen_scores.append(slot_scores[slot][idx])
            else:
                chosen_scores.append(0.0)

        # 슬롯별 상위 키워드 및 점수 정보 추가
        top_keywords_dict = {f"{slot}_top_keywords": slot_keywords[slot] for slot in slot_names}
        top_scores_dict = {f"{slot}_top_scores": slot_scores[slot] for slot in slot_names}

        recommendations.append({
            "template": template,
            "title_candidate": title_candidate.strip(),
            "slot_mapping": dict(zip(slot_names, combo)),
            "slot_scores": dict(zip(slot_names, chosen_scores)),
            **top_keywords_dict,
            **top_scores_dict
        })

# DataFrame으로 정리
df_recommendations = pd.DataFrame(recommendations)

# NaN 안전 처리
def replace_nan_with_list(x):
    if isinstance(x, list):
        return x
    elif pd.isna(x):
        return []
    else:
        return x

for col in df_recommendations.columns:
    if col.endswith("_top_keywords") or col.endswith("_top_scores"):
        df_recommendations[col] = df_recommendations[col].apply(replace_nan_with_list)

# 중복된 title_candidate 제거
df_recommendations = df_recommendations.drop_duplicates(subset=['title_candidate']).reset_index(drop=True)

df_recommendations = df_recommendations.drop(columns=['{SITUATION}_top_keywords', '{TOPIC}_top_keywords', '{SITUATION}_top_scores', '{TOPIC}_top_scores'])
df_recommendations
```
