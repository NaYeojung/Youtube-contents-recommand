## ì œëª© ì „ì²˜ë¦¬

```python
import requests

url = "https://raw.githubusercontent.com/stopwords-iso/stopwords-ko/master/stopwords-ko.txt"
STOPWORDS = requests.get(url).text.splitlines()
```

```python
import re, numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer

# 0) ì…ë ¥ ê°€ì •
# df_all: ìœ„ ë‹¨ê³„ì—ì„œ ë§Œë“  ê²°í•© DF (í•„ìˆ˜: 'title')
# í•„ìš”í•˜ë©´ ì¤‘ë³µ/ê²°ì¸¡ ì œê±°
df_all = df_all.copy()
df_all = df_all.dropna(subset=["title"])
df_all = df_all.drop_duplicates(subset=["title"]).reset_index(drop=True)

# 1) ê°„ë‹¨ í† í¬ë‚˜ì´ì € + ë¶ˆìš©ì–´
def tokenize_ko(text: str):
    text = re.sub(r"[^0-9ê°€-í£A-Za-z]+", " ", str(text))
    toks = text.lower().split()
    return [t for t in toks if len(t) > 1]

def analyzer(doc): return [t for t in tokenize_ko(doc) if t not in STOPWORDS]

df_all["tokens"] = df_all["title"].apply(analyzer)
print(df_all[["title", "tokens"]].head(10))
```

```python
# 2) TF-IDF ì í•©
vectorizer = TfidfVectorizer(
    analyzer=analyzer,
    ngram_range=(1,2),  # ë‹¨ì–´+ë°”ì´ê·¸ë¨
    min_df=2            # 2ê°œ ì´ìƒ ë¬¸ì„œì—ì„œ ë“±ì¥í•œ í† í°ë§Œ
)
X = vectorizer.fit_transform(df_all["title"].astype(str))
vocab = np.array(vectorizer.get_feature_names_out())

# 3) ì „ì²´ ìƒìœ„ í‚¤ì›Œë“œ(top-K)
top_k = 100
scores = np.asarray(X.sum(axis=0)).ravel()
order = scores.argsort()[::-1][:top_k]
top_keywords = list(zip(vocab[order], scores[order]))

print("[ì „ì²´ ìƒìœ„ í‚¤ì›Œë“œ]")
for k, s in top_keywords:
    print(f"{k}\t{s:.3f}")
```

## í‚¤ì›Œë“œ ì¶”ì¶œ

```python
# =========================
# YouTube ì œëª© í‚¤ì›Œë“œ/í…œí”Œë¦¿ ì¶”ì¶œ (í•œ ì…€ ì™„ê²°)
# =========================

# 0) Imports
import os, re, requests, numpy as np, pandas as pd
from tqdm import tqdm

# 1) ì…ë ¥ ì •ë¦¬ (title í•„ìˆ˜)
assert "df_all" in globals(), "df_all DataFrameì´ ë©”ëª¨ë¦¬ì— ìˆì–´ì•¼ í•©ë‹ˆë‹¤."
assert "title" in df_all.columns, "df_allì— 'title' ì»¬ëŸ¼ì´ í•„ìš”í•©ë‹ˆë‹¤."
df_all = df_all.copy().dropna(subset=["title"]).drop_duplicates(subset=["title"]).reset_index(drop=True)

# 2) í•œêµ­ì–´ ë¶ˆìš©ì–´ ì„¸íŠ¸ ë¡œë“œ(ê¹ƒí—ˆë¸Œ + í´ë°±) â€” ë°˜ë“œì‹œ set í˜•íƒœë¡œ!
try:
    url = "https://raw.githubusercontent.com/stopwords-iso/stopwords-ko/master/stopwords-ko.txt"
    STOPWORDS = set(w.strip().lower() for w in requests.get(url, timeout=10).text.splitlines() if w.strip())
except Exception:
    STOPWORDS = {"ê·¸ë¦¬ê³ ","ê·¸ëŸ¬ë‚˜","í•˜ì§€ë§Œ","ê·¸ë˜ì„œ","ë˜í•œ","í•©ë‹ˆë‹¤","í–ˆë‹¤","í•˜ëŠ”","í•˜ê²Œ","ë˜ì–´","ë˜ëŠ”",
                 "ì—ê²Œ","ì—ì„œ","ìœ¼ë¡œ","ë³´ë‹¤","ê°™ì€","ì´ë²ˆ","ì˜¤ëŠ˜","ê·¸ê²ƒ","ì €ê²ƒ","ì´ê²ƒ","ì¶”ì²œ","í›„ê¸°","ë¦¬ë·°",
                 "ëª¨ìŒ","ì´ì •ë¦¬","ê¿€íŒ","ê°€ì´ë“œ","ë°©ë²•","ë¹„ë²•","ë¸Œì´ë¡œê·¸","ì˜ìƒ","ìµœì‹ ","í•„ìˆ˜","ì£¼ì˜","ì‹¤í™”","ì´ìŠˆ"}
# placeholderë¥˜ ì¶”ê°€
STOPWORDS |= {"url","email","hashtag","num","eng"}

# 3) í† í¬ë‚˜ì´ì €(ê°„ë‹¨) + ë¶ˆìš©ì–´ ì ìš©
def tokenize_ko(text: str):
    text = re.sub(r"https?://\S+|www\.\S+", " url ", str(text))
    text = re.sub(r"[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}", " email ", text)
    text = re.sub(r"#\S+", " hashtag ", text)
    text = re.sub(r"\d+([.,]\d+)*", " num ", text)
    text = re.sub(r"[^0-9ê°€-í£A-Za-z]+", " ", text)
    toks = text.lower().split()
    return [t for t in toks if len(t) > 1 and t not in STOPWORDS]


df_all["tokens"] = df_all["tokens"].apply(
    lambda xs: [t.lower() for t in (xs or []) if isinstance(t, str) and t.strip() and t.lower() not in STOPWORDS]
)

print(df_all[["title","tokens"]].head(10))
```

```python
# 4) TF-IDF í‚¤ì›Œë“œ (í”„ë¦¬í† í° ì…ë ¥ ì „ìš© â€” stop_words ì „ë‹¬ ê¸ˆì§€)
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import normalize

def tfidf_keywords_from_tokens(tokens_list, top_k=6, ngram_range=(1,2), min_df=2, max_df=0.95):
    vec = TfidfVectorizer(
        analyzer=lambda x: x,   # ì´ë¯¸ í† í° ë¦¬ìŠ¤íŠ¸
        tokenizer=None,
        preprocessor=None,
        token_pattern=None,
        lowercase=False,
        ngram_range=ngram_range,
        min_df=min_df,
        max_df=max_df,
        stop_words=None         # âš ï¸ ë°˜ë“œì‹œ None (ê²½ê³  ë°©ì§€)
    )
    X = vec.fit_transform(tokens_list)
    X = normalize(X, norm="l2", copy=False)
    vocab = vec.get_feature_names_out()

    results = []
    for i in tqdm(range(X.shape[0]), desc="TF-IDF keywords"):
        row = X.getrow(i)
        if row.nnz == 0:
            results.append([]); continue
        ord_idx = row.indices[row.data.argsort()[::-1][:max(top_k*6, top_k)]]
        cand = [(vocab[j], float(row[0, j])) for j in ord_idx]
        cleaned, added = [], set()
        for token, score in cand:
            if len(token) < 2: continue
            if token in added: continue
            cleaned.append((token, score)); added.add(token)
            if len(cleaned) >= top_k: break
        results.append(cleaned)
    return results, vec

tokens_list = df_all["tokens"].tolist()
tfidf_res, tfidf_vec = tfidf_keywords_from_tokens(tokens_list, top_k=6, ngram_range=(1,2), min_df=2, max_df=0.95)

# 5) KeyBERT í‚¤ì›Œë“œ (í•œêµ­ì–´ ë¶ˆìš©ì–´ë¥¼ tokenizerì—ì„œ ì¬ì ìš©í•´ ê²½ê³  ì—†ì´ ì¼ê´€ ì²˜ë¦¬)
def keybert_from_tokens(tokens_list, top_k=6, model_name="snunlp/KR-SBERT-V40K-klueNLI-augSTS"):
    try:
        from keybert import KeyBERT
        from sentence_transformers import SentenceTransformer
        from sklearn.feature_extraction.text import CountVectorizer
    except Exception:
        return [[] for _ in tokens_list]

    # SBERT ë¡œë“œ(í´ë°± í¬í•¨)
    try:
        model = SentenceTransformer(model_name)
    except Exception:
        model = SentenceTransformer("paraphrase-multilingual-MiniLM-L12-v2")
    kw_model = KeyBERT(model=model)

    # ì»¤ìŠ¤í…€ tokenizer: split â†’ STOPWORDS ì œê±° (í•œêµ­ì–´ ë¦¬ìŠ¤íŠ¸ ê°•ì œ ì ìš©)
    def _tok(s: str):
        words = s.split()
        return [w for w in words if w not in STOPWORDS and len(w) > 1]

    vec = CountVectorizer(
        tokenizer=_tok, preprocessor=None, token_pattern=None,
        lowercase=False, ngram_range=(1,3), stop_words=None, min_df=1
    )

    results = []
    for toks in tqdm(tokens_list, desc="KeyBERT keywords"):
        text = " ".join(toks) if isinstance(toks, list) else str(toks)
        kws = kw_model.extract_keywords(
            text,
            vectorizer=vec,        # í›„ë³´ ìƒì„±ì— ìš°ë¦¬ tokenizer ì‚¬ìš©
            use_maxsum=True,
            nr_candidates=20,
            top_n=top_k,
        )
        results.append([(k, float(s)) for k, s in kws])
    return results

keybert_res = keybert_from_tokens(tokens_list, top_k=6)


df_all = df_all.copy()
df_all["tfidf_keywords"] = tfidf_res
df_all["keybert_keywords"] = keybert_res
```

```python
# =========================
# í…œí”Œë¦¿ ìƒì„± (word-level, í‘œë©´í˜• ë³´ì¡´)
# ìˆœì„œ: Kkmaë¡œ ê³ ìœ ëª…ì‚¬/ìˆ«ì íŒë‹¨ â†’ PERSON(ë³´ìˆ˜) â†’ NUM â†’ ENG â†’ LOC â†’ NOUN â†’ ê´„í˜¸{X}
# =========================
import re, unicodedata
from tqdm import tqdm
from konlpy.tag import Kkma

kkma = Kkma()

# (ìˆìœ¼ë©´ ì‚¬ìš©; ì—†ìœ¼ë©´ ë¹ˆ ì„¸íŠ¸)
keep_tokens = {*(keep_tokens if 'keep_tokens' in globals() else set())}
STOPWORDS  = {*(STOPWORDS  if 'STOPWORDS'  in globals() else set())}

# ------ ìœ í‹¸: ìœ ë‹ˆì½”ë“œ ì•ˆì •í™” & í† í°í™”(í‘œë©´í˜• ìœ ì§€) ------
_SURROGATE_RE = re.compile(r"[\ud800-\udfff]")
_ZERO_WIDTH_RE = re.compile(r"[\u200B-\u200F\uFEFF]")
_EMOJI_SUPP_RE = re.compile(r"[\U00010000-\U0010FFFF]")

def clean_text(s: str) -> str:
    s = unicodedata.normalize("NFC", str(s))
    s = _ZERO_WIDTH_RE.sub(" ", s)
    s = _SURROGATE_RE.sub(" ", s)
    s = _EMOJI_SUPP_RE.sub(" ", s)
    return s

# ë‹¨ì–´/êµ¬ë‘ì /ê³µë°±ì„ ë¶„ë¦¬í•˜ë˜, ì›ë˜ í‘œë©´í˜•ê³¼ ìˆœì„œë¥¼ ê·¸ëŒ€ë¡œ ë³´ì¡´
# ex) ["ë…¸í™ì² ", " ", "ë˜", " ", "í•œë²ˆ", " ", "ë†€ë¼ê²Œ", " ", "ë§Œë“ ", " ", "ì¶©ê²©ì ì¸", " ", "ë™ìœ ëŸ½", " ", "ë¬¼ê°€", "(", "ë¶ˆê°€ë¦¬ì•„", ")"]
WORD_SPLIT = re.compile(r"([^\wê°€-í£]+)")  # 'ë‹¨ì–´'ê°€ ì•„ë‹Œ ë¶€ë¶„(ìŠ¤í˜ì´ìŠ¤/êµ¬ë‘ì )ì„ ë”°ë¡œ ìº¡ì²˜

def split_preserve(text: str):
    parts = []
    for chunk in WORD_SPLIT.split(text):
        if chunk is None or chunk == "":
            continue
        parts.append(chunk)
    return parts

def is_word(token: str) -> bool:
    # í•œê¸€, ì˜ë¬¸, ìˆ«ì í¬í•¨ëœ ë¬¶ìŒì„ 'ë‹¨ì–´'ë¡œ ê°„ì£¼ (ê³µë°±/êµ¬ë‘ì ì€ False)
    return bool(re.search(r"[A-Za-z0-9ê°€-í£]", token))

# ------ íŒ¨í„´ & ì‚¬ì „ ------
COMMON_SURNAMES = {
    "ê¹€","ì´","ë°•","ìµœ","ì •","ê°•","ì¡°","ìœ¤","ì¥","ì„","í•œ","ì˜¤","ì„œ","ì‹ ","ê¶Œ","í™©","ì•ˆ","ì†¡","ì „","í™","ìœ ","ê³ ","ë¬¸",
    "ì–‘","ì†","ë°°","ë°±","í—ˆ","ë‚¨","ì‹¬","ë…¸","í•˜","ê³½","ì„±","ì°¨","ì£¼","ìš°","êµ¬","ë¯¼","ë¥˜","ë‚˜","ì§„","ì§€","ì—„","ì±„","ì›",
    "ì²œ","ë°©","ê³µ","í˜„","í•¨","ë³€","ì—¼","ì—¬","ì¶”","ë„","ì†Œ","ì„","ì„ ","ì„¤","ë§ˆ","ê¸¸","ì—°","ìœ„","í‘œ","ëª…","ê¸°","ë°˜","ë¼",
    "ì™•","ê¸ˆ","ì˜¥","ìœ¡","ì¸","ë§¹","ì œ","ëª¨","ë‚¨ê¶","íƒ","êµ­","ì–´"
}
ENG_WORD  = re.compile(r"^[A-Za-z]{2,}$")           # ìˆœìˆ˜ ì˜ë¬¸(2+)
ENG_CAP   = re.compile(r"^[A-Z][a-z]{1,20}$")       # ì˜ë¬¸ ì´ë¦„ í† í° ëª¨ì–‘
NUM_FULL  = re.compile(r"^\d+(?:[.,]\d+)*$")        # ìˆœìˆ˜ ìˆ«ì
NUM_INNER = re.compile(r"\d+")                      # í˜¼í•© ìˆ«ì í¬í•¨(4K, 1080p ë“±)
BRACKETS_RE = re.compile(r"(\([^)]+\)|\[[^\]]+\])") # (), [] ë§Œ {X} í›„ë³´

def mask_brackets_with_X(s: str, numeric_ratio=0.6):
    def _f(m):
        inner = m.group(0)[1:-1]
        if not inner.strip(): return m.group(0)
        digits = len(re.findall(r"\d", inner))
        nonalpha = len(re.findall(r"[^ê°€-í£A-Za-z]", inner))
        total = max(1, len(inner))
        ratio = max(digits, nonalpha) / total
        if ratio >= numeric_ratio:
            return m.group(0)[0] + " {X} " + m.group(0)[-1]
        return m.group(0)
    return BRACKETS_RE.sub(_f, s)

# ------ í•œêµ­ ì „êµ­ ì§€ì—­ ì‚¬ì „ ------
# â‘  17ê°œ ì‹œÂ·ë„ (ê´‘ì—­ë‹¨ìœ„)
LOC_KO = {
    # íŠ¹ë³„ì‹œ/ê´‘ì—­ì‹œ/íŠ¹ë³„ìì¹˜ì‹œ
    "ì„œìš¸","ë¶€ì‚°","ëŒ€êµ¬","ì¸ì²œ","ê´‘ì£¼","ëŒ€ì „","ìš¸ì‚°","ì„¸ì¢…",
    # ë„/íŠ¹ë³„ìì¹˜ë„
    "ê²½ê¸°","ê°•ì›","ì¶©ë¶","ì¶©ë‚¨","ì „ë¶","ì „ë‚¨","ê²½ë¶","ê²½ë‚¨","ì œì£¼","ì œì£¼ë„",
    # ì‹œê¸‰(ì£¼ìš” ëŒ€ë„ì‹œ ë° ê´‘ì—­ì‹œ ì‚°í•˜ êµ¬Â·ì‹œ ì¼ë¶€)
    "ìˆ˜ì›","ì„±ë‚¨","ìš©ì¸","ê³ ì–‘","ê³¼ì²œ","ê´‘ëª…","ì•ˆì–‘","ë¶€ì²œ","ì•ˆì‚°","í‰íƒ","ì˜ì •ë¶€","ê¹€í¬","íŒŒì£¼","ë‚¨ì–‘ì£¼","í•˜ë‚¨","êµ¬ë¦¬",
    "ì‹œí¥","êµ°í¬","ì˜ì™•","ì˜¤ì‚°","ì´ì²œ","ì—¬ì£¼","ì–‘í‰","ë™ë‘ì²œ","ê°€í‰","ì—°ì²œ","í¬ì²œ",
    "ì¶˜ì²œ","ì›ì£¼","ê°•ë¦‰","ë™í•´","ì†ì´ˆ","ì‚¼ì²™","íƒœë°±","í™ì²œ","íš¡ì„±","ì˜ì›”","í‰ì°½","ì •ì„ ","ì² ì›","í™”ì²œ","ì–‘êµ¬","ì¸ì œ","ê³ ì„±","ì–‘ì–‘",
    "ì²­ì£¼","ì¶©ì£¼","ì œì²œ","ë³´ì€","ì˜¥ì²œ","ì˜ë™","ì¦í‰","ì§„ì²œ","ê´´ì‚°","ìŒì„±","ë‹¨ì–‘",
    "ì²œì•ˆ","ê³µì£¼","ë³´ë ¹","ì•„ì‚°","ì„œì‚°","ë…¼ì‚°","ê³„ë£¡","ë‹¹ì§„","ê¸ˆì‚°","ë¶€ì—¬","ì„œì²œ","ì²­ì–‘","í™ì„±","ì˜ˆì‚°","íƒœì•ˆ",
    "ì „ì£¼","êµ°ì‚°","ìµì‚°","ì •ì","ë‚¨ì›","ê¹€ì œ","ì™„ì£¼","ì§„ì•ˆ","ë¬´ì£¼","ì¥ìˆ˜","ì„ì‹¤","ìˆœì°½","ê³ ì°½","ë¶€ì•ˆ",
    "ëª©í¬","ì—¬ìˆ˜","ìˆœì²œ","ë‚˜ì£¼","ê´‘ì–‘","ë‹´ì–‘","ê³¡ì„±","êµ¬ë¡€","ê³ í¥","ë³´ì„±","í™”ìˆœ","ì¥í¥","ê°•ì§„","í•´ë‚¨","ì˜ì•”","ë¬´ì•ˆ","í•¨í‰","ì˜ê´‘","ì¥ì„±","ì™„ë„","ì§„ë„","ì‹ ì•ˆ",
    "í¬í•­","ê²½ì£¼","ê¹€ì²œ","ì•ˆë™","êµ¬ë¯¸","ì˜ì£¼","ì˜ì²œ","ìƒì£¼","ë¬¸ê²½","ê²½ì‚°","êµ°ìœ„","ì˜ì„±","ì²­ì†¡","ì˜ì–‘","ì˜ë•","ì²­ë„","ê³ ë ¹","ì„±ì£¼","ì¹ ê³¡","ì˜ˆì²œ","ë´‰í™”","ìš¸ì§„","ìš¸ë¦‰",
    "ì°½ì›","ì§„ì£¼","í†µì˜","ì‚¬ì²œ","ê¹€í•´","ë°€ì–‘","ê±°ì œ","ì–‘ì‚°","ì˜ë ¹","í•¨ì•ˆ","ì°½ë…•","ê³ ì„±","ë‚¨í•´","í•˜ë™","ì‚°ì²­","í•¨ì–‘","ê±°ì°½","í•©ì²œ",
    "ì œì£¼ì‹œ","ì„œê·€í¬ì‹œ",
    # ê´‘ì—­ì‹œ ì‚°í•˜ êµ¬Â·êµ°(ëŒ€í‘œ/ìì£¼ ë“±ì¥)
    "í•´ìš´ëŒ€","ìˆ˜ì˜","ë‚¨êµ¬","ë¶êµ¬","ë™êµ¬","ì„œêµ¬","ì¤‘êµ¬","ì˜ë„","ê¸ˆì •","ì—°ì œ","ë™ë˜","ì‚¬í•˜","ì‚¬ìƒ","ê°•ì„œ","ê¸°ì¥",
    "ìˆ˜ì„±","ë‹¬ì„œ","ë‹¬ì„±","ë™êµ¬","ì„œêµ¬","ë‚¨êµ¬","ë¶êµ¬","ì¤‘êµ¬",
    "ë¯¸ì¶”í™€","ì—°ìˆ˜","ë‚¨ë™","ë¶€í‰","ê³„ì–‘","ê°•í™”","ì˜¹ì§„",
    "ë‚¨êµ¬","ë¶êµ¬","ê´‘ì‚°",
    "ìœ ì„±","ëŒ€ë•",
    "ë‚¨êµ¬","ë™êµ¬","ì¤‘êµ¬","ë¶êµ¬","ìš¸ì£¼",
    # ê¸°íƒ€ ë§ì´ ì“°ëŠ” ê´€ê´‘ë„ì‹œ/ì§€ì—­ëª…
    "ì—¬ì˜ë„","ê°•ë‚¨","í™ëŒ€","ëª…ë™","ë™ëŒ€ë¬¸","ë‚¨ëŒ€ë¬¸","ì´íƒœì›","ì„œì´Œ","ë¶ì´Œ","ê°€í‰","ì–‘ì–‘","ì†ì´ˆ","ì—¬ìˆ˜","ì œì£¼ë„","í•œë¼ì‚°","ì„±ì‚°","ì¤‘ë¬¸","ì• ì›”","í˜‘ì¬"
}

# ------ PERSON íƒì§€(ë³´ìˆ˜): ë‹¨ì–´ ì‹œí€€ìŠ¤ ê¸°ì¤€ ------
def detect_person_spans(words, poss):
    """
    words: ë‹¨ì–´/êµ¬ë‘ì /ê³µë°± ì „ì²´ ì‹œí€€ìŠ¤
    poss:  ê° 'ë‹¨ì–´' í† í°ì— ëŒ€í•œ Kkma.pos ê²°ê³¼ì˜ ìš”ì•½(label per word).
           ë¹„ë‹¨ì–´(ê³µë°±/êµ¬ë‘ì )ì—ëŠ” None.
    ë°˜í™˜: set[int]  -> PERSONìœ¼ë¡œ ì¹˜í™˜í•  'ë‹¨ì–´' ì¸ë±ìŠ¤ ì§‘í•© (ë³‘í•© ì‹œ ì²« ë‹¨ì–´ë§Œ ì¹˜í™˜)
    ê·œì¹™(ë³´ìˆ˜):
      - í•œê¸€ ì´ë¦„: ì—°ì† ë‹¨ì–´ê°€ (1ì/2ì + 2ì) ë“± 2~4ìŒì ˆì´ê³ , ê° ë‹¨ì–´ì˜ ì£¼ìš” í’ˆì‚¬ê°€ NNPì´ë©°, í•©ì¹œ ë¬¸ìì—´ì˜ ì²« ê¸€ìê°€ ì„±ì”¨
      - ë‹¨ì¼ ë‹¨ì–´ í•œê¸€ 2~3ì, ì£¼ìš” í’ˆì‚¬ NNP, ì²« ê¸€ì ì„±ì”¨
      - ì˜ë¬¸ ì´ë¦„: ENG_CAP ë‹¨ì–´ 2ê°œ ì—°ì† (ë‘˜ ë‹¤ ì£¼ìš” í’ˆì‚¬ NNPë¡œ ë‚˜ì˜¤ì§€ ì•Šì•„ë„ í—ˆìš©)
    """
    person_heads = set()
    def is_nnp(i_word):
        labs = poss[i_word] or set()
        return ("NNP" in labs)
    def has_korean_only(s):
        return bool(re.fullmatch(r"[ê°€-í£]+", s))
    # 1) ì˜ë¬¸ í’€ë„¤ì„
    for i in range(len(words)-1):
        if not (is_word(words[i]) and is_word(words[i+1])): 
            continue
        a, b = words[i], words[i+1]
        if ENG_CAP.match(a) and ENG_CAP.match(b):
            person_heads.add(i)  # i ìœ„ì¹˜ì— {PERSON}ì„ ë„£ê³ , i+1ì€ ìŠ¤í‚µ
    # 2) í•œê¸€ ì´ë¦„ (ê³µë°± í¬í•¨)
    n = len(words); i = 0
    while i < n:
        if not is_word(words[i]): 
            i += 1; continue
        w = words[i]
        if has_korean_only(w):
            # ë‹¨ì¼ ë‹¨ì–´ 2~3ì, ì„±ì”¨ ì‹œì‘ & NNP
            if 2 <= len(w) <= 3 and w[0] in COMMON_SURNAMES and is_nnp(i):
                person_heads.add(i); i += 1; continue
            # ê³µë°± í¬í•¨ 2~4ì
            j = i
            merged = w
            while j+1 < n and is_word(words[j+1]) and has_korean_only(words[j+1]) and len(merged) < 4:
                merged += words[j+1]; j += 1
            if 2 <= len(merged) <= 4 and merged[0] in COMMON_SURNAMES:
                ok = True
                for k in range(i, j+1):
                    labs = poss[k] or set()
                    if "NNP" not in labs: ok = False; break
                if ok:
                    person_heads.add(i)
                    i = j + 1
                    continue
        i += 1
    return person_heads

# ------ ê° ë‹¨ì–´ì— ëŒ€í•´ Kkma.pos ìš”ì•½ ------
def pos_labels_per_word(token: str):
    """
    í•œ ë‹¨ì–´ì— ëŒ€í•´ Kkma.pos ì‹¤í–‰ í›„, ì£¼ìš” í’ˆì‚¬ set ë°˜í™˜.
    í‘œë©´í˜• ìœ ì§€ ëª©ì ì´ë¼ ë‹¨ì–´ ë‹¨ìœ„ë¡œë§Œ íƒœê¹…/íŒë‹¨í•˜ê³  ì¹˜í™˜ì€ ë‹¨ì–´ ì „ì²´ì— ëŒ€í•´ ìˆ˜í–‰.
    """
    labs = set()
    try:
        for t, p in kkma.pos(token):
            labs.add(p)
    except Exception:
        pass
    return labs

# ------ ë©”ì¸: í•œ ì œëª© â†’ í…œí”Œë¦¿ ------
def to_template_surface(title: str,
                        min_noun_len=3,
                        max_mask_ratio=0.55,
                        max_person_per_title=2,
                        bracket_numeric_ratio=0.6):
    s0 = clean_text(title)
    # 0) URL/EMAIL/HASHTAG ì„ ì¹˜í™˜(í‘œë©´í˜• ë³´ì¡´)
    s = re.sub(r"https?://\S+|www\.\S+", " {URL} ", s0)
    s = re.sub(r"[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}", " {EMAIL} ", s)
    s = re.sub(r"#\S+", " {HASHTAG} ", s)

    # 1) ë‹¨ì–´/êµ¬ë‘ì /ê³µë°± ë¶„ë¦¬ (í‘œë©´í˜• ìœ ì§€)
    parts = split_preserve(s)

    # 2) ê° 'ë‹¨ì–´' íŒŒíŠ¸ì— ëŒ€í•´ Kkma.pos ìš”ì•½ ë¼ë²¨ ìˆ˜ì§‘
    labels = [pos_labels_per_word(w) if is_word(w) and w not in {"{URL}","{EMAIL}","{HASHTAG}"} else None
              for w in parts]

    # 3) PERSON í—¤ë“œ ì¸ë±ìŠ¤ íƒì§€(ë³´ìˆ˜)
    person_heads = detect_person_spans(parts, labels)

    # 4) ë‹¨ì–´ë³„ ì¹˜í™˜: PERSON â†’ NUM â†’ ENG â†’ LOC â†’ NOUN (í‘œë©´í˜• ìì²´ëŠ” ë³€í˜•í•˜ì§€ ì•ŠìŒ)
    out = []
    person_used = 0
    masked = 0

    for i, w in enumerate(parts):
        # ë¹„ë‹¨ì–´(ê³µë°±/êµ¬ë‘ì ) ë˜ëŠ” ë¯¸ë¦¬ ì¹˜í™˜ëœ í”Œë ˆì´ìŠ¤í™€ë”ëŠ” ê·¸ëŒ€ë¡œ
        if not is_word(w) or w in {"{URL}","{EMAIL}","{HASHTAG}"}:
            out.append(w); continue

        low = w.lower()

        # 4-1) PERSON (ë³´ìˆ˜, ì—°ì† spanì˜ ì²« ë‹¨ì–´ì—ë§Œ ì¹˜í™˜)
        if i in person_heads and person_used < max_person_per_title:
            out.append("{PERSON}")
            person_used += 1
            masked += 1
            continue

        # 4-2) ìˆ«ì: ìˆœìˆ˜/í˜¼í•©
        labs = labels[i] or set()
        if NUM_FULL.match(w) or ("SN" in labs) or ("NR" in labs):
            out.append("{NUM}"); masked += 1; continue
        if NUM_INNER.search(w) and not ENG_WORD.match(w):
            out.append(NUM_INNER.sub("{NUM}", w)); masked += 1; continue

        # 4-3) ìˆœìˆ˜ ENG
        if ENG_WORD.match(w):
            out.append("{ENG}"); masked += 1; continue

        # 4-4) LOC (ì „êµ­ ì§€ì—­ ì‚¬ì „)  â† ì¶”ê°€ëœ ë¶€ë¶„
        if w in LOC_KO:
            out.append("{LOC}"); masked += 1; continue

        # 4-5) ìƒìœ„ í‚¤ì›Œë“œ/ë¶ˆìš©ì–´ ë³´í˜¸
        if low in keep_tokens or low in STOPWORDS:
            out.append(w); continue

        # 4-6) ì¼ë°˜ëª…ì‚¬/ê³ ìœ ëª…ì‚¬: í‘œë©´í˜•ì„ {NOUN}ìœ¼ë¡œë§Œ ì¹˜í™˜(í˜•íƒœ ë³€í˜• ì—†ìŒ)
        if "NNP" in labs:
            out.append("{NOUN}"); masked += 1; continue
        if "NNG" in labs and len(w) >= min_noun_len:
            out.append("{NOUN}"); masked += 1; continue

        # 4-7) ê·¸ ì™¸ í’ˆì‚¬(ë™ì‚¬/í˜•ìš©ì‚¬/ì¡°ì‚¬ ë“±)ëŠ” ì›í˜• ìœ ì§€
        out.append(w)

    # 5) ê³¼ë§ˆìŠ¤í‚¹ ì™„í™”(ë¹„ìœ¨ ì´ˆê³¼ ì‹œ ì¼ë¶€ {NOUN}/{PERSON}/{LOC} ì œê±°)
    total_tokens = max(1, sum(1 for w in parts if is_word(w)))
    if masked / total_tokens > max_mask_ratio:
        need = int(masked - max_mask_ratio * total_tokens)
        for j, t in enumerate(out):
            if need <= 0: break
            if t in ("{PERSON}","{NOUN}","{LOC}"):
                out[j] = ""  # ì œê±° â†’ ê³µë°± ì •ë¦¬ë¡œ ìì—° ì¶•ì•½
                need -= 1

    # 6) ì¡°ë¦½ & í›„ì²˜ë¦¬ (ì—°ì† í”Œë ˆì´ìŠ¤í™€ë” ì¶•ì•½ + ê´„í˜¸ {X})
    out_str = "".join(out)  # í‘œë©´í˜• ë³´ì¡´: partsë¥¼ ê·¸ëŒ€ë¡œ ì´ì–´ë¶™ì„
    out_str = re.sub(r"(?:\{PERSON\}\s*){2,}", "{PERSON} ", out_str)
    out_str = re.sub(r"(?:\{NUM\}\s*){2,}", "{NUM} ", out_str)
    out_str = re.sub(r"(?:\{NOUN\}\s*){2,}", "{NOUN} ", out_str)
    out_str = re.sub(r"(?:\{LOC\}\s*){2,}", "{LOC} ", out_str)
    out_str = re.sub(r"\s+", " ", out_str).strip()

    # ê´„í˜¸ X ì²˜ë¦¬ (ìˆ«ì/ê¸°í˜¸ ìœ„ì£¼ì¼ ë•Œë§Œ)
    out_str = mask_brackets_with_X(out_str, numeric_ratio=bracket_numeric_ratio)

    return out_str

# ---------- ì‹¤í–‰ ----------
templates = [to_template_surface(t,
                                 min_noun_len=3,
                                 max_mask_ratio=0.55,
                                 max_person_per_title=2,
                                 bracket_numeric_ratio=0.6)
             for t in tqdm(df_all["title"].astype(str), desc="Templates (surface-safe + LOC)")]

df_all["title_template"] = templates

template_summary = (
    df_all["title_template"].fillna("").replace("", None).dropna()
      .value_counts().reset_index().rename(columns={"index":"template","title_template":"count"})
)
print(template_summary.head(15))
```

```python
# =========================
# í…œí”Œë¦¿ ìƒì„± (word-level, í‘œë©´í˜• ë³´ì¡´)
# ìˆœì„œ: Kkmaë¡œ ê³ ìœ ëª…ì‚¬/ìˆ«ì íŒë‹¨ â†’ PERSON â†’ NUM(ì¡°ì‚¬ ë³´ì¡´) â†’ ENG â†’ LOC(ì¡°ì‚¬ ë³´ì¡´) â†’ NOUN(ì¡°ì‚¬ ë³´ì¡´) â†’ ê´„í˜¸{X}
# =========================
import re, unicodedata
from tqdm import tqdm
from konlpy.tag import Kkma

kkma = Kkma()

# (ìˆìœ¼ë©´ ì‚¬ìš©; ì—†ìœ¼ë©´ ë¹ˆ ì„¸íŠ¸)
keep_tokens = {*(keep_tokens if 'keep_tokens' in globals() else set())}
STOPWORDS  = {*(STOPWORDS  if 'STOPWORDS'  in globals() else set())}

# ------ ìœ í‹¸: ìœ ë‹ˆì½”ë“œ ì•ˆì •í™” & í† í°í™”(í‘œë©´í˜• ë³´ì¡´) ------
_SURROGATE_RE = re.compile(r"[\ud800-\udfff]")
_ZERO_WIDTH_RE = re.compile(r"[\u200B-\u200F\uFEFF]")
_EMOJI_SUPP_RE = re.compile(r"[\U00010000-\U0010FFFF]")

def clean_text(s: str) -> str:
    s = unicodedata.normalize("NFC", str(s))
    s = _ZERO_WIDTH_RE.sub(" ", s)
    s = _SURROGATE_RE.sub(" ", s)
    s = _EMOJI_SUPP_RE.sub(" ", s)
    return s

# ë‹¨ì–´/êµ¬ë‘ì /ê³µë°± ë¶„ë¦¬ (í‘œë©´í˜• ë³´ì¡´)
WORD_SPLIT = re.compile(r"([^\wê°€-í£]+)")

def split_preserve(text: str):
    parts = []
    for chunk in WORD_SPLIT.split(text):
        if chunk is None or chunk == "":
            continue
        parts.append(chunk)
    return parts

def is_word(token: str) -> bool:
    return bool(re.search(r"[A-Za-z0-9ê°€-í£]", token))

# ------ ì¡°ì‚¬ ì‚¬ì „(ê¸´ ê²ƒ ìš°ì„ ) & ë¶„ë¦¬ ------
JOSA_SUFFIXES = [
    "ìœ¼ë¡œì¨","ìœ¼ë¡œì„œ","ìœ¼ë¡œë¶€í„°","ë¡œë¶€í„°",
    "ì´ì—ˆìŠµë‹ˆë‹¤","ì˜€ìŠµë‹ˆë‹¤","ì´ì—ˆë‹¤","ì˜€ë‹¤","ì´ì—ˆë‹¤ê°€","ì˜€ë‹¤ê°€",
    "ì´ë¼ê³ ëŠ”","ë¼ê³ ëŠ”","ì´ë¼ê³ ","ë¼ê³ ",
    "ì—ê²Œì„œ","ì—ê²Œë§Œ","ì—ê²Œë„","ì—ê²Œ","ê»˜ì„œ","ê»˜",
    "ì—ì„œëŠ”","ì—ì„œ","ì—ëŠ”ë§Œ","ì—ëŠ”ë„","ì—ëŠ”","ì—ëŠ”ë„","ì—ëŠ”",
    "ìœ¼ë¡œëŠ”","ë¡œëŠ”","ìœ¼ë¡œ","ë¡œ",
    "ì´ë¼ë©´","ë¼ë©´","ì´ë¼ì„œ","ë¼ì„œ","ì´ì§€ë§Œ","ì§€ë§Œ","ì´ê±°ë‚˜","ê±°ë‚˜",
    "ì´ë©°ì„œ","ë©´ì„œ","ì´ë©°","ë©°","ì´ê³ ","ê³ ",
    "ê³¼ì˜","ì™€ì˜","ê³¼ëŠ”","ì™€ëŠ”","ê³¼ë„","ì™€ë„","ê³¼ë§Œ","ì™€ë§Œ","ê³¼","ì™€","ë‘ì€","ë‘ë„","ë‘ë§Œ","ë‘",
    "ì€","ëŠ”","ì´","ê°€","ì„","ë¥¼","ë„","ë§Œ","ë¿","ê¹Œì§€","ë¶€í„°","ë§ˆì €","ì¡°ì°¨","ì´ë‚˜","ë‚˜","ì˜"
]
JOSA_SUFFIXES = sorted(list({j for j in JOSA_SUFFIXES if re.fullmatch(r"[ê°€-í£]+", j)}), key=len, reverse=True)

def split_stem_josa(word: str):
    """ì–´ì ˆ ëì—ì„œ ì¡°ì‚¬ í•˜ë‚˜ë¥¼ ë–¼ì–´ (stem, josa)ë¡œ ë°˜í™˜. ì—†ìœ¼ë©´ (word, "")"""
    for suf in JOSA_SUFFIXES:
        if word.endswith(suf) and len(word) > len(suf):
            return word[:-len(suf)], suf
    return word, ""

# ------ íŒ¨í„´ & ì‚¬ì „ ------
COMMON_SURNAMES = {
    "ê¹€","ì´","ë°•","ìµœ","ì •","ê°•","ì¡°","ìœ¤","ì¥","ì„","í•œ","ì˜¤","ì„œ","ì‹ ","ê¶Œ","í™©","ì•ˆ","ì†¡","ì „","í™","ìœ ","ê³ ","ë¬¸",
    "ì–‘","ì†","ë°°","ë°±","í—ˆ","ë‚¨","ì‹¬","ë…¸","í•˜","ê³½","ì„±","ì°¨","ì£¼","ìš°","êµ¬","ë¯¼","ë¥˜","ë‚˜","ì§„","ì§€","ì—„","ì±„","ì›",
    "ì²œ","ë°©","ê³µ","í˜„","í•¨","ë³€","ì—¼","ì—¬","ì¶”","ë„","ì†Œ","ì„","ì„ ","ì„¤","ë§ˆ","ê¸¸","ì—°","ìœ„","í‘œ","ëª…","ê¸°","ë°˜","ë¼",
    "ì™•","ê¸ˆ","ì˜¥","ìœ¡","ì¸","ë§¹","ì œ","ëª¨","ë‚¨ê¶","íƒ","êµ­","ì–´"
}
ENG_WORD  = re.compile(r"^[A-Za-z]{2,}$")
ENG_CAP   = re.compile(r"^[A-Z][a-z]{1,20}$")
NUM_FULL  = re.compile(r"^\d+(?:[.,]\d+)*$")
NUM_INNER = re.compile(r"\d+")
BRACKETS_RE = re.compile(r"(\([^)]+\)|\[[^\]]+\])")

def mask_brackets_with_X(s: str, numeric_ratio=0.6):
    def _f(m):
        inner = m.group(0)[1:-1]
        if not inner.strip(): return m.group(0)
        digits = len(re.findall(r"\d", inner))
        nonalpha = len(re.findall(r"[^ê°€-í£A-Za-z]", inner))
        total = max(1, len(inner))
        ratio = max(digits, nonalpha) / total
        if ratio >= numeric_ratio:
            return m.group(0)[0] + " {X} " + m.group(0)[-1]
        return m.group(0)
    return BRACKETS_RE.sub(_f, s)

# ------ ì „êµ­ ì§€ì—­ ì‚¬ì „ (ì‹œÂ·ë„/ì‹œÂ·êµ°Â·êµ¬ + ì£¼ìš” ìƒê¶Œ ì¼ë¶€) ------
LOC_KO = {
    # 17ê°œ ì‹œÂ·ë„
    "ì„œìš¸","ë¶€ì‚°","ëŒ€êµ¬","ì¸ì²œ","ê´‘ì£¼","ëŒ€ì „","ìš¸ì‚°","ì„¸ì¢…",
    "ê²½ê¸°","ê°•ì›","ì¶©ë¶","ì¶©ë‚¨","ì „ë¶","ì „ë‚¨","ê²½ë¶","ê²½ë‚¨","ì œì£¼","ì œì£¼ë„",
    # ìˆ˜ë„ê¶Œ/ê²½ê¸°
    "ìˆ˜ì›","ì„±ë‚¨","ìš©ì¸","ê³ ì–‘","ë¶€ì²œ","ì•ˆì–‘","ì•ˆì‚°","í‰íƒ","ì˜ì •ë¶€","ê¹€í¬","íŒŒì£¼","ë‚¨ì–‘ì£¼","í•˜ë‚¨","êµ¬ë¦¬","ì‹œí¥",
    "ê´‘ëª…","êµ°í¬","ì˜ì™•","ì˜¤ì‚°","ê³¼ì²œ","ì´ì²œ","ì—¬ì£¼","ì–‘í‰","ë™ë‘ì²œ","ê°€í‰","ì—°ì²œ","í¬ì²œ",
    # ê°•ì›
    "ì¶˜ì²œ","ì›ì£¼","ê°•ë¦‰","ë™í•´","ì†ì´ˆ","ì‚¼ì²™","íƒœë°±","í™ì²œ","íš¡ì„±","ì˜ì›”","í‰ì°½","ì •ì„ ","ì² ì›","í™”ì²œ","ì–‘êµ¬","ì¸ì œ","ê³ ì„±","ì–‘ì–‘",
    # ì¶©ë¶
    "ì²­ì£¼","ì¶©ì£¼","ì œì²œ","ë³´ì€","ì˜¥ì²œ","ì˜ë™","ì¦í‰","ì§„ì²œ","ê´´ì‚°","ìŒì„±","ë‹¨ì–‘",
    # ì¶©ë‚¨
    "ì²œì•ˆ","ê³µì£¼","ë³´ë ¹","ì•„ì‚°","ì„œì‚°","ë…¼ì‚°","ê³„ë£¡","ë‹¹ì§„","ê¸ˆì‚°","ë¶€ì—¬","ì„œì²œ","ì²­ì–‘","í™ì„±","ì˜ˆì‚°","íƒœì•ˆ",
    # ì „ë¶
    "ì „ì£¼","êµ°ì‚°","ìµì‚°","ì •ì","ë‚¨ì›","ê¹€ì œ","ì™„ì£¼","ì§„ì•ˆ","ë¬´ì£¼","ì¥ìˆ˜","ì„ì‹¤","ìˆœì°½","ê³ ì°½","ë¶€ì•ˆ",
    # ì „ë‚¨
    "ëª©í¬","ì—¬ìˆ˜","ìˆœì²œ","ë‚˜ì£¼","ê´‘ì–‘","ë‹´ì–‘","ê³¡ì„±","êµ¬ë¡€","ê³ í¥","ë³´ì„±","í™”ìˆœ","ì¥í¥","ê°•ì§„","í•´ë‚¨","ì˜ì•”","ë¬´ì•ˆ","í•¨í‰","ì˜ê´‘","ì¥ì„±","ì™„ë„","ì§„ë„","ì‹ ì•ˆ",
    # ê²½ë¶
    "í¬í•­","ê²½ì£¼","ê¹€ì²œ","ì•ˆë™","êµ¬ë¯¸","ì˜ì£¼","ì˜ì²œ","ìƒì£¼","ë¬¸ê²½","ê²½ì‚°","êµ°ìœ„","ì˜ì„±","ì²­ì†¡","ì˜ì–‘","ì˜ë•","ì²­ë„","ê³ ë ¹","ì„±ì£¼","ì¹ ê³¡","ì˜ˆì²œ","ë´‰í™”","ìš¸ì§„","ìš¸ë¦‰",
    # ê²½ë‚¨
    "ì°½ì›","ì§„ì£¼","í†µì˜","ì‚¬ì²œ","ê¹€í•´","ë°€ì–‘","ê±°ì œ","ì–‘ì‚°","ì˜ë ¹","í•¨ì•ˆ","ì°½ë…•","ê³ ì„±","ë‚¨í•´","í•˜ë™","ì‚°ì²­","í•¨ì–‘","ê±°ì°½","í•©ì²œ",
    # ì œì£¼
    "ì œì£¼ì‹œ","ì„œê·€í¬ì‹œ",
    # ê´‘ì—­ì‹œ í•˜ìœ„ êµ¬(ìì£¼ ì“°ëŠ” ê³³ ìœ„ì£¼)
    "í•´ìš´ëŒ€","ìˆ˜ì˜","ë‚¨êµ¬","ë¶êµ¬","ë™êµ¬","ì„œêµ¬","ì¤‘êµ¬","ì˜ë„","ê¸ˆì •","ì—°ì œ","ë™ë˜","ì‚¬í•˜","ì‚¬ìƒ","ê°•ì„œ","ê¸°ì¥",
    "ìˆ˜ì„±","ë‹¬ì„œ","ë‹¬ì„±",
    "ë¯¸ì¶”í™€","ì—°ìˆ˜","ë‚¨ë™","ë¶€í‰","ê³„ì–‘","ê°•í™”","ì˜¹ì§„",
    "ê´‘ì‚°","ìœ ì„±","ëŒ€ë•","ìš¸ì£¼",
    # ìœ ëª… ìƒê¶Œ/ê´€ê´‘ì§€
    "ê°•ë‚¨","ì—¬ì˜ë„","í™ëŒ€","ëª…ë™","ë™ëŒ€ë¬¸","ë‚¨ëŒ€ë¬¸","ì´íƒœì›","ì„œì´Œ","ë¶ì´Œ",
    "ì œì£¼ê³µí•­","ì„±ì‚°","ì¤‘ë¬¸","ì• ì›”","í˜‘ì¬"
}

# ------ PERSON íƒì§€(ë³´ìˆ˜) ------
def detect_person_spans(words, poss):
    person_heads = set()

    def is_nnp(i_word):
        labs = poss[i_word] or set()
        return ("NNP" in labs)
    def has_korean_only(s):
        return bool(re.fullmatch(r"[ê°€-í£]+", s))

    # ì˜ë¬¸ í’€ë„¤ì„
    for i in range(len(words)-1):
        if not (is_word(words[i]) and is_word(words[i+1])): 
            continue
        a, b = words[i], words[i+1]
        if ENG_CAP.match(a) and ENG_CAP.match(b):
            person_heads.add(i)

    # í•œê¸€ ì´ë¦„
    n = len(words); i = 0
    while i < n:
        if not is_word(words[i]): 
            i += 1; continue
        w = words[i]
        if has_korean_only(w):
            # ë‹¨ì¼ 2~3ì NNP + ì„±ì”¨
            if 2 <= len(w) <= 3 and w[0] in COMMON_SURNAMES and is_nnp(i):
                person_heads.add(i); i += 1; continue
            # ê³µë°± í¬í•¨ 2~4ì, ëª¨ë‘ NNP
            j = i
            merged = w
            while j+1 < n and is_word(words[j+1]) and has_korean_only(words[j+1]) and len(merged) < 4:
                merged += words[j+1]; j += 1
            if 2 <= len(merged) <= 4 and merged[0] in COMMON_SURNAMES:
                ok = True
                for k in range(i, j+1):
                    labs = poss[k] or set()
                    if "NNP" not in labs: ok = False; break
                if ok:
                    person_heads.add(i)
                    i = j + 1
                    continue
        i += 1
    return person_heads

# ------ ê° ë‹¨ì–´ì— ëŒ€í•´ Kkma.pos ìš”ì•½ ------
def pos_labels_per_word(token: str):
    labs = set()
    try:
        for t, p in kkma.pos(token):
            labs.add(p)
    except Exception:
        pass
    return labs

# ------ ë©”ì¸: í•œ ì œëª© â†’ í…œí”Œë¦¿ ------
def to_template_surface(title: str,
                        min_noun_len=3,
                        max_mask_ratio=0.55,    # (ì°¸ê³ ê°’) ì§€ê¸ˆì€ ì‚­ì œ/ë³µì› ì•ˆ í•¨
                        max_person_per_title=2,
                        bracket_numeric_ratio=0.6):
    s0 = clean_text(title)
    # 0) URL/EMAIL/HASHTAG ì„ ì¹˜í™˜
    s = re.sub(r"https?://\S+|www\.\S+", " {URL} ", s0)
    s = re.sub(r"[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}", " {EMAIL} ", s)
    s = re.sub(r"#\S+", " {HASHTAG} ", s)

    # 1) ë¶„í•´(í‘œë©´í˜• ë³´ì¡´)
    parts = split_preserve(s)

    # 2) POS ìš”ì•½
    labels = [pos_labels_per_word(w) if is_word(w) and w not in {"{URL}","{EMAIL}","{HASHTAG}"} else None
              for w in parts]

    # 3) PERSON í—¤ë“œ íƒì§€
    person_heads = detect_person_spans(parts, labels)

    # 4) ì¹˜í™˜: PERSON â†’ NUM(ì¡°ì‚¬ ë³´ì¡´) â†’ ENG â†’ LOC(ì¡°ì‚¬ ë³´ì¡´) â†’ NOUN(ì¡°ì‚¬ ë³´ì¡´)
    out = []
    person_used = 0

    for i, w in enumerate(parts):
        if not is_word(w) or w in {"{URL}","{EMAIL}","{HASHTAG}"}:
            out.append(w); continue

        low = w.lower()
        labs = labels[i] or set()

        # ì¡°ì‚¬ ë¶„ë¦¬(ì´í•˜ ë³´ì¡´ ëŒ€ìƒì—ì„œ ì‚¬ìš©)
        stem, josa = split_stem_josa(w)

        # 4-1) PERSON (í—¤ë“œë§Œ, ì¡°ì‚¬ ë³´ì¡´ ì—¬ë¶€ëŠ” ì„ íƒ â€” í•„ìš” ì‹œ ì•„ë˜ì²˜ëŸ¼ ë¶™ì´ê¸°)
        if i in person_heads and person_used < max_person_per_title:
            # out.append("{PERSON}"+josa)  # â† PERSONì—ë„ ì¡°ì‚¬ ë³´ì¡´ì„ ì›í•˜ë©´ ì´ ì¤„ ì‚¬ìš©
            out.append("{PERSON}")         # â† ê¸°ë³¸: PERSONì€ ì¡°ì‚¬ ë¯¸ë³´ì¡´(ìš”ì²­ ì‹œ ìœ„ë¡œ êµì²´)
            person_used += 1
            continue

        # 4-2) NUM â€” ì¡°ì‚¬ ë³´ì¡´
        if NUM_FULL.match(stem) or ("SN" in labs) or ("NR" in labs):
            out.append("{NUM}"+josa); continue
        if NUM_INNER.search(stem) and not ENG_WORD.match(stem):
            out.append(NUM_INNER.sub("{NUM}", stem) + josa); continue

        # 4-3) ENG (ì¡°ì‚¬ ë¶™ëŠ” ì¼€ì´ìŠ¤ ë“œë¬¼ì§€ë§Œ, í•„ìš” ì‹œ ë³´ì¡´í•´ë„ ë¨)
        if ENG_WORD.match(stem):
            # out.append("{ENG}"+josa)   # ENGì—ë„ ì¡°ì‚¬ ë³´ì¡´ ì›í•˜ë©´ ì‚¬ìš©
            out.append("{ENG}"); continue

        # 4-4) LOC â€” ì¡°ì‚¬ ë³´ì¡´
        if stem in LOC_KO or w in LOC_KO:
            if stem in LOC_KO and josa:
                out.append("{LOC}"+josa); continue
            if w in LOC_KO:
                out.append("{LOC}"); continue

        # 4-5) ìƒìœ„ í‚¤ì›Œë“œ/ë¶ˆìš©ì–´ ë³´í˜¸
        if low in keep_tokens or low in STOPWORDS:
            out.append(w); continue

        # 4-6) NOUN â€” ì¡°ì‚¬ ë³´ì¡´
        if "NNP" in labs:
            out.append("{NOUN}"+josa); continue
        if "NNG" in labs and len(stem) >= min_noun_len:
            out.append("{NOUN}"+josa); continue

        # 4-7) ê·¸ ì™¸ëŠ” ì›í˜• ìœ ì§€
        out.append(w)

    # 5) ê³¼ë§ˆìŠ¤í‚¹ ì™„í™” â€” ì‚­ì œ ì—†ìŒ(ê·¸ëŒ€ë¡œ ìœ ì§€)
    #    í•„ìš” ì‹œ 'ë„ˆë¬´ íƒœê·¸ê°€ ë§ë‹¤'ê³  ëŠê»´ì§€ë©´, ì´í›„ ë‹¨ê³„ì—ì„œ ë¦¬ë­í¬ë§Œ ê³ ë ¤í•˜ì„¸ìš”.

    # 6) ì¡°ë¦½ & í›„ì²˜ë¦¬
    out_str = "".join(out)
    out_str = re.sub(r"(?:\{PERSON\}\s*){2,}", "{PERSON} ", out_str)
    out_str = re.sub(r"(?:\{NUM\}\s*){2,}", "{NUM} ", out_str)
    out_str = re.sub(r"(?:\{NOUN\}\s*){2,}", "{NOUN} ", out_str)
    out_str = re.sub(r"(?:\{LOC\}\s*){2,}", "{LOC} ", out_str)
    out_str = re.sub(r"\s+", " ", out_str).strip()
    out_str = mask_brackets_with_X(out_str, numeric_ratio=bracket_numeric_ratio)
    return out_str

# ---------- ì‹¤í–‰ ----------
templates = [to_template_surface(t,
                                 min_noun_len=3,
                                 max_mask_ratio=0.55,
                                 max_person_per_title=2,
                                 bracket_numeric_ratio=0.6)
             for t in tqdm(df_all["title"].astype(str), desc="Templates (surface-safe + LOC/NUM/NOUN josa)")]

df_all["title_template"] = templates

# ìš”ì•½
template_summary = (
    df_all["title_template"].fillna("").replace("", None).dropna()
      .value_counts().reset_index().rename(columns={"index":"template","title_template":"count"})
)
print(template_summary.head(15))
```

```python
# =========================
# í…œí”Œë¦¿ ìƒì„± (word-level, í‘œë©´í˜• ë³´ì¡´)
# ìˆœì„œ: LOC(ì¡°ì‚¬ ë³´ì¡´) â†’ PERSON â†’ NUM(ì¡°ì‚¬ ë³´ì¡´) â†’ ENG â†’ NOUN(ì¡°ì‚¬ ë³´ì¡´) â†’ ê´„í˜¸{X}
# =========================
import re, unicodedata
from tqdm import tqdm
from konlpy.tag import Kkma

kkma = Kkma()

# (ìˆìœ¼ë©´ ì‚¬ìš©; ì—†ìœ¼ë©´ ë¹ˆ ì„¸íŠ¸)
keep_tokens = {*(keep_tokens if 'keep_tokens' in globals() else set())}
STOPWORDS  = {*(STOPWORDS  if 'STOPWORDS'  in globals() else set())}

# ------ ìœ í‹¸: ìœ ë‹ˆì½”ë“œ ì•ˆì •í™” & í† í°í™”(í‘œë©´í˜• ë³´ì¡´) ------
_SURROGATE_RE = re.compile(r"[\ud800-\udfff]")
_ZERO_WIDTH_RE = re.compile(r"[\u200B-\u200F\uFEFF]")
_EMOJI_SUPP_RE = re.compile(r"[\U00010000-\U0010FFFF]")

def clean_text(s: str) -> str:
    s = unicodedata.normalize("NFC", str(s))
    s = _ZERO_WIDTH_RE.sub(" ", s)
    s = _SURROGATE_RE.sub(" ", s)
    s = _EMOJI_SUPP_RE.sub(" ", s)
    return s

# ë‹¨ì–´/êµ¬ë‘ì /ê³µë°± ë¶„ë¦¬ (í‘œë©´í˜• ë³´ì¡´)
WORD_SPLIT = re.compile(r"([^\wê°€-í£]+)")

def split_preserve(text: str):
    parts = []
    for chunk in WORD_SPLIT.split(text):
        if chunk:
            parts.append(chunk)
    return parts

def is_word(token: str) -> bool:
    return bool(re.search(r"[A-Za-z0-9ê°€-í£]", token))

# ------ ì¡°ì‚¬ ì‚¬ì „(ê¸´ ê²ƒ ìš°ì„ ) & ë¶„ë¦¬ ------
JOSA_SUFFIXES = [
    "ìœ¼ë¡œì¨","ìœ¼ë¡œì„œ","ìœ¼ë¡œë¶€í„°","ë¡œë¶€í„°",
    "ì´ì—ˆìŠµë‹ˆë‹¤","ì˜€ìŠµë‹ˆë‹¤","ì´ì—ˆë‹¤","ì˜€ë‹¤","ì´ì—ˆë‹¤ê°€","ì˜€ë‹¤ê°€",
    "ì´ë¼ê³ ëŠ”","ë¼ê³ ëŠ”","ì´ë¼ê³ ","ë¼ê³ ",
    "ì—ê²Œì„œ","ì—ê²Œë§Œ","ì—ê²Œë„","ì—ê²Œ","ê»˜ì„œ","ê»˜",
    "ì—ì„œëŠ”","ì—ì„œ","ì—ëŠ”ë§Œ","ì—ëŠ”ë„","ì—ëŠ”","ì—ëŠ”ë„","ì—ëŠ”",
    "ìœ¼ë¡œëŠ”","ë¡œëŠ”","ìœ¼ë¡œ","ë¡œ",
    "ì´ë¼ë©´","ë¼ë©´","ì´ë¼ì„œ","ë¼ì„œ","ì´ì§€ë§Œ","ì§€ë§Œ","ì´ê±°ë‚˜","ê±°ë‚˜",
    "ì´ë©°ì„œ","ë©´ì„œ","ì´ë©°","ë©°","ì´ê³ ","ê³ ",
    "ê³¼ì˜","ì™€ì˜","ê³¼ëŠ”","ì™€ëŠ”","ê³¼ë„","ì™€ë„","ê³¼ë§Œ","ì™€ë§Œ","ê³¼","ì™€","ë‘ì€","ë‘ë„","ë‘ë§Œ","ë‘",
    "ì€","ëŠ”","ì´","ê°€","ì„","ë¥¼","ë„","ë§Œ","ë¿","ê¹Œì§€","ë¶€í„°","ë§ˆì €","ì¡°ì°¨","ì´ë‚˜","ë‚˜","ì˜"
]
JOSA_SUFFIXES = sorted({j for j in JOSA_SUFFIXES if re.fullmatch(r"[ê°€-í£]+", j)}, key=len, reverse=True)

def split_stem_josa(word: str):
    """ì–´ì ˆ ëì˜ ì¡°ì‚¬ 1ê°œë¥¼ ë–¼ì–´ (stem, josa)ë¡œ ë°˜í™˜. ì¡°ì‚¬ ì—†ìœ¼ë©´ (word, "")"""
    for suf in JOSA_SUFFIXES:
        if word.endswith(suf) and len(word) > len(suf):
            return word[:-len(suf)], suf
    return word, ""

# ------ íŒ¨í„´ & ì‚¬ì „ ------
COMMON_SURNAMES = {
    "ê¹€","ì´","ë°•","ìµœ","ì •","ê°•","ì¡°","ìœ¤","ì¥","ì„","í•œ","ì˜¤","ì„œ","ì‹ ","ê¶Œ","í™©","ì•ˆ","ì†¡","ì „","í™","ìœ ","ê³ ","ë¬¸",
    "ì–‘","ì†","ë°°","ë°±","í—ˆ","ë‚¨","ì‹¬","ë…¸","í•˜","ê³½","ì„±","ì°¨","ì£¼","ìš°","êµ¬","ë¯¼","ë¥˜","ë‚˜","ì§„","ì§€","ì—„","ì±„","ì›",
    "ì²œ","ë°©","ê³µ","í˜„","í•¨","ë³€","ì—¼","ì—¬","ì¶”","ë„","ì†Œ","ì„","ì„ ","ì„¤","ë§ˆ","ê¸¸","ì—°","ìœ„","í‘œ","ëª…","ê¸°","ë°˜","ë¼",
    "ì™•","ê¸ˆ","ì˜¥","ìœ¡","ì¸","ë§¹","ì œ","ëª¨","ë‚¨ê¶","íƒ","êµ­","ì–´"
}
ENG_WORD  = re.compile(r"^[A-Za-z]{2,}$")
ENG_CAP   = re.compile(r"^[A-Z][a-z]{1,20}$")
NUM_FULL  = re.compile(r"^\d+(?:[.,]\d+)*$")
NUM_INNER = re.compile(r"\d+")
BRACKETS_RE = re.compile(r"(\([^)]+\)|\[[^\]]+\])")

def mask_brackets_with_X(s: str, numeric_ratio=0.6):
    def _f(m):
        inner = m.group(0)[1:-1]
        if not inner.strip(): return m.group(0)
        digits = len(re.findall(r"\d", inner))
        nonalpha = len(re.findall(r"[^ê°€-í£A-Za-z]", inner))
        total = max(1, len(inner))
        ratio = max(digits, nonalpha) / total
        if ratio >= numeric_ratio:
            return m.group(0)[0] + " {X} " + m.group(0)[-1]
        return m.group(0)
    return BRACKETS_RE.sub(_f, s)

# ------ í•œêµ­ ì „êµ­ ì§€ì—­ ì‚¬ì „ (ì‹œÂ·ë„/ì‹œÂ·êµ°Â·êµ¬ + ì£¼ìš” ìƒê¶Œ ì¼ë¶€) ------
LOC_KO = {
    "ì„œìš¸","ë¶€ì‚°","ëŒ€êµ¬","ì¸ì²œ","ê´‘ì£¼","ëŒ€ì „","ìš¸ì‚°","ì„¸ì¢…",
    "ê²½ê¸°","ê°•ì›","ì¶©ë¶","ì¶©ë‚¨","ì „ë¶","ì „ë‚¨","ê²½ë¶","ê²½ë‚¨","ì œì£¼","ì œì£¼ë„",
    "ìˆ˜ì›","ì„±ë‚¨","ìš©ì¸","ê³ ì–‘","ë¶€ì²œ","ì•ˆì–‘","ì•ˆì‚°","í‰íƒ","ì˜ì •ë¶€","ê¹€í¬","íŒŒì£¼","ë‚¨ì–‘ì£¼","í•˜ë‚¨","êµ¬ë¦¬","ì‹œí¥",
    "ê´‘ëª…","êµ°í¬","ì˜ì™•","ì˜¤ì‚°","ê³¼ì²œ","ì´ì²œ","ì—¬ì£¼","ì–‘í‰","ë™ë‘ì²œ","ê°€í‰","ì—°ì²œ","í¬ì²œ",
    "ì¶˜ì²œ","ì›ì£¼","ê°•ë¦‰","ë™í•´","ì†ì´ˆ","ì‚¼ì²™","íƒœë°±","í™ì²œ","íš¡ì„±","ì˜ì›”","í‰ì°½","ì •ì„ ","ì² ì›","í™”ì²œ","ì–‘êµ¬","ì¸ì œ","ê³ ì„±","ì–‘ì–‘",
    "ì²­ì£¼","ì¶©ì£¼","ì œì²œ","ë³´ì€","ì˜¥ì²œ","ì˜ë™","ì¦í‰","ì§„ì²œ","ê´´ì‚°","ìŒì„±","ë‹¨ì–‘",
    "ì²œì•ˆ","ê³µì£¼","ë³´ë ¹","ì•„ì‚°","ì„œì‚°","ë…¼ì‚°","ê³„ë£¡","ë‹¹ì§„","ê¸ˆì‚°","ë¶€ì—¬","ì„œì²œ","ì²­ì–‘","í™ì„±","ì˜ˆì‚°","íƒœì•ˆ",
    "ì „ì£¼","êµ°ì‚°","ìµì‚°","ì •ì","ë‚¨ì›","ê¹€ì œ","ì™„ì£¼","ì§„ì•ˆ","ë¬´ì£¼","ì¥ìˆ˜","ì„ì‹¤","ìˆœì°½","ê³ ì°½","ë¶€ì•ˆ",
    "ëª©í¬","ì—¬ìˆ˜","ìˆœì²œ","ë‚˜ì£¼","ê´‘ì–‘","ë‹´ì–‘","ê³¡ì„±","êµ¬ë¡€","ê³ í¥","ë³´ì„±","í™”ìˆœ","ì¥í¥","ê°•ì§„","í•´ë‚¨","ì˜ì•”","ë¬´ì•ˆ","í•¨í‰","ì˜ê´‘","ì¥ì„±","ì™„ë„","ì§„ë„","ì‹ ì•ˆ",
    "í¬í•­","ê²½ì£¼","ê¹€ì²œ","ì•ˆë™","êµ¬ë¯¸","ì˜ì£¼","ì˜ì²œ","ìƒì£¼","ë¬¸ê²½","ê²½ì‚°","êµ°ìœ„","ì˜ì„±","ì²­ì†¡","ì˜ì–‘","ì˜ë•","ì²­ë„","ê³ ë ¹","ì„±ì£¼","ì¹ ê³¡","ì˜ˆì²œ","ë´‰í™”","ìš¸ì§„","ìš¸ë¦‰",
    "ì°½ì›","ì§„ì£¼","í†µì˜","ì‚¬ì²œ","ê¹€í•´","ë°€ì–‘","ê±°ì œ","ì–‘ì‚°","ì˜ë ¹","í•¨ì•ˆ","ì°½ë…•","ê³ ì„±","ë‚¨í•´","í•˜ë™","ì‚°ì²­","í•¨ì–‘","ê±°ì°½","í•©ì²œ",
    "ì œì£¼ì‹œ","ì„œê·€í¬ì‹œ",
    
    "í•´ìš´ëŒ€","ìˆ˜ì˜","ë‚¨êµ¬","ë¶êµ¬","ë™êµ¬","ì„œêµ¬","ì¤‘êµ¬","ì˜ë„","ê¸ˆì •","ì—°ì œ","ë™ë˜","ì‚¬í•˜","ì‚¬ìƒ","ê°•ì„œ","ê¸°ì¥",
    "ìˆ˜ì„±","ë‹¬ì„œ","ë‹¬ì„±",
    "ë¯¸ì¶”í™€","ì—°ìˆ˜","ë‚¨ë™","ë¶€í‰","ê³„ì–‘","ê°•í™”","ì˜¹ì§„",
    "ê´‘ì‚°","ìœ ì„±","ëŒ€ë•","ìš¸ì£¼",
    "ê°•ë‚¨","ì—¬ì˜ë„","í™ëŒ€","ëª…ë™","ë™ëŒ€ë¬¸","ë‚¨ëŒ€ë¬¸","ì´íƒœì›","ì„œì´Œ","ë¶ì´Œ",
    "ì œì£¼ê³µí•­","ì„±ì‚°","ì¤‘ë¬¸","ì• ì›”","í˜‘ì¬"
}

# ------ ê° ë‹¨ì–´ì— ëŒ€í•´ Kkma.pos ìš”ì•½ ------
def pos_labels_per_word(token: str):
    labs = set()
    try:
        for t, p in kkma.pos(token):
            labs.add(p)
    except Exception:
        pass
    return labs

# ------ PERSON íƒì§€(ë³´ìˆ˜) â€” LOC ì œì™¸ ë°˜ì˜ ------
def detect_person_spans(words, poss, loc_set=None):
    """
    loc_set: LOC_KO ì§‘í•©ì„ ë„˜ê¸°ë©´, í•´ë‹¹ ë‹¨ì–´/ë³‘í•©ì–´ê°€ LOCì¼ ê²½ìš° PERSON í›„ë³´ì—ì„œ ì œì™¸
    """
    if loc_set is None:
        loc_set = set()
    person_heads = set()

    def is_nnp(i_word):
        labs = poss[i_word] or set()
        return ("NNP" in labs)

    def has_korean_only(s):
        return bool(re.fullmatch(r"[ê°€-í£]+", s))

    # ì˜ë¬¸ í’€ë„¤ì„ (LOC ì œì™¸)
    for i in range(len(words)-1):
        if not (is_word(words[i]) and is_word(words[i+1])): 
            continue
        a, b = words[i], words[i+1]
        if (a in loc_set) or (b in loc_set):
            continue
        if ENG_CAP.match(a) and ENG_CAP.match(b):
            person_heads.add(i)

    # í•œê¸€ ì´ë¦„ (LOC ì œì™¸)
    n = len(words); i = 0
    while i < n:
        if not is_word(words[i]): 
            i += 1; continue
        w = words[i]
        if w in loc_set:           # ì§€ì—­ ë‹¨ì–´ëŠ” ìŠ¤í‚µ
            i += 1; continue
        if has_korean_only(w):
            # ë‹¨ì¼ 2~3ì NNP + ì„±ì”¨
            if 2 <= len(w) <= 3 and w[0] in COMMON_SURNAMES and is_nnp(i) and (w not in loc_set):
                person_heads.add(i); i += 1; continue
            # ê³µë°± í¬í•¨ 2~4ì, ëª¨ë‘ NNP
            j = i
            merged = w
            while j+1 < n and is_word(words[j+1]) and has_korean_only(words[j+1]) and len(merged) < 4:
                merged += words[j+1]; j += 1
            if (merged not in loc_set) and 2 <= len(merged) <= 4 and merged[0] in COMMON_SURNAMES:
                ok = True
                for k in range(i, j+1):
                    labs = poss[k] or set()
                    if "NNP" not in labs: ok = False; break
                if ok:
                    person_heads.add(i)
                    i = j + 1
                    continue
        i += 1
    return person_heads

# ------ ë©”ì¸: í•œ ì œëª© â†’ í…œí”Œë¦¿ ------
def to_template_surface(title: str,
                        min_noun_len=3,
                        max_mask_ratio=0.55,    # í˜„ì¬ëŠ” ì‚­ì œ/ë³µì› ì—†ìŒ
                        max_person_per_title=2,
                        bracket_numeric_ratio=0.6):
    s0 = clean_text(title)
    # 0) URL/EMAIL/HASHTAG ì„ ì¹˜í™˜
    s = re.sub(r"https?://\S+|www\.\S+", " {URL} ", s0)
    s = re.sub(r"[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}", " {EMAIL} ", s)
    s = re.sub(r"#\S+", " {HASHTAG} ", s)

    # 1) ë¶„í•´(í‘œë©´í˜• ë³´ì¡´)
    parts = split_preserve(s)

    # 2) POS ìš”ì•½
    labels = [pos_labels_per_word(w) if is_word(w) and w not in {"{URL}","{EMAIL}","{HASHTAG}"} else None
              for w in parts]

    # 3) PERSON í—¤ë“œ ì¸ë±ìŠ¤ (LOC ì œì™¸ ë°˜ì˜)
    person_heads = detect_person_spans(parts, labels, loc_set=LOC_KO)

    # 4) ì¹˜í™˜: LOC(ì¡°ì‚¬ ë³´ì¡´) â†’ PERSON â†’ NUM(ì¡°ì‚¬ ë³´ì¡´) â†’ ENG â†’ NOUN(ì¡°ì‚¬ ë³´ì¡´)
    out = []
    person_used = 0

    for i, w in enumerate(parts):
        if not is_word(w) or w in {"{URL}","{EMAIL}","{HASHTAG}"}:
            out.append(w); continue

        low = w.lower()
        labs = labels[i] or set()

        # ì¡°ì‚¬ ë¶„ë¦¬ (ì¡°ì‚¬ ë³´ì¡´ìš©)
        stem, josa = split_stem_josa(w)

        # 4-1) LOC â€” ë¨¼ì € ì²˜ë¦¬ (ì¡°ì‚¬ ë³´ì¡´)
        if stem in LOC_KO or w in LOC_KO:
            if stem in LOC_KO and josa:
                out.append("{LOC}"+josa); continue
            if w in LOC_KO:
                out.append("{LOC}"); continue

        # 4-2) PERSON â€” LOC íƒœê¹… ë’¤ ìˆ˜í–‰
        if i in person_heads and person_used < max_person_per_title:
            # PERSONì—ë„ ì¡°ì‚¬ ë³´ì¡´ì„ ì›í•˜ë©´ ì•„ë˜ ì£¼ì„ í•´ì œ
            # out.append("{PERSON}"+josa); person_used += 1; continue
            out.append("{PERSON}"); person_used += 1; continue

        # 4-3) NUM â€” ì¡°ì‚¬ ë³´ì¡´
        if NUM_FULL.match(stem) or ("SN" in labs) or ("NR" in labs):
            out.append("{NUM}"+josa); continue
        if NUM_INNER.search(stem) and not ENG_WORD.match(stem):
            out.append(NUM_INNER.sub("{NUM}", stem) + josa); continue

        # 4-4) ENG (í•„ìš” ì‹œ ì¡°ì‚¬ ë³´ì¡´ìœ¼ë¡œ ë°”ê¿€ ìˆ˜ ìˆìŒ)
        if ENG_WORD.match(stem):
            # out.append("{ENG}"+josa); continue
            out.append("{ENG}"); continue

        # 4-5) ìƒìœ„ í‚¤ì›Œë“œ/ë¶ˆìš©ì–´ ë³´í˜¸
        if low in keep_tokens or low in STOPWORDS:
            out.append(w); continue

        # 4-6) NOUN â€” ì¡°ì‚¬ ë³´ì¡´
        if "NNP" in labs:
            out.append("{NOUN}"+josa); continue
        if "NNG" in labs and len(stem) >= min_noun_len:
            out.append("{NOUN}"+josa); continue

        # 4-7) ê·¸ ì™¸(ë™ì‚¬/í˜•ìš©ì‚¬/ë¶€ì‚¬/ê°íƒ„ì‚¬/ì¡°ì‚¬ ë‹¨ë… ë“±)ëŠ” ì›í˜• ìœ ì§€
        out.append(w)

    # 5) ê³¼ë§ˆìŠ¤í‚¹ ì™„í™” â€” ì‚­ì œ ì—†ìŒ(í‘œë©´í˜• ë³´ì¡´)
    out_str = "".join(out)
    out_str = re.sub(r"(?:\{PERSON\}\s*){2,}", "{PERSON} ", out_str)
    out_str = re.sub(r"(?:\{NUM\}\s*){2,}", "{NUM} ", out_str)
    out_str = re.sub(r"(?:\{NOUN\}\s*){2,}", "{NOUN} ", out_str)
    out_str = re.sub(r"(?:\{LOC\}\s*){2,}", "{LOC} ", out_str)
    out_str = re.sub(r"\s+", " ", out_str).strip()

    # 6) ê´„í˜¸ {X} ì²˜ë¦¬ (ìˆ«ì/ê¸°í˜¸ ìœ„ì£¼ì¼ ë•Œë§Œ)
    out_str = mask_brackets_with_X(out_str, numeric_ratio=bracket_numeric_ratio)
    return out_str

# ---------- ì‹¤í–‰ ----------
templates = [to_template_surface(
                t,
                min_noun_len=3,
                max_mask_ratio=0.55,
                max_person_per_title=2,
                bracket_numeric_ratio=0.6
            ) for t in tqdm(df_all["title"].astype(str), desc="Templates (LOC â†’ PERSON â†’ NUM â†’ ENG â†’ NOUN)")]

df_all["title_template"] = templates

# ìš”ì•½ í™•ì¸
template_summary = (
    df_all["title_template"].fillna("").replace("", None).dropna()
    .value_counts().reset_index().rename(columns={"index":"template","title_template":"count"})
)
print(template_summary.head(15))
```
